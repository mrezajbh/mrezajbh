```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

class AdsorptionColumnSimulation:
    def __init__(self):
        # Column properties
        self.d = 4.0  # diameter of column [m]
        self.h = 7.0  # height of column [m]
        self.Qmh = 681400  # Gas mass flow rate [kg/h]
        self.Qm = self.Qmh / 3600  # Gas mass flow [kg/s]
        self.channeling = 0
        self.Channeling = 100 * self.channeling  # % part of gas that's channeling

        # Discretization
        self.ny = 240  # number of height sections
        self.dy = self.h / self.ny  # step size in y-direction
        self.tmax = 60 * 100  # [s] Total duration of simulation
        self.dt = 0.025  # [s] size of timestep
        self.nt = int(self.tmax / self.dt)  # total number of time steps

        # Column areas
        self.cA = (np.pi * self.d**2) / 4  # cross-sectional area of column
        self.aA = 0.62 * self.cA  # Area the adsorbent is filling
        self.fA = self.cA - self.aA  # free cross-sectional area for gas flow

        # Adsorbent properties
        self.hvap = 4187  # heat of vaporization [KJ/kg]
        self.Cpa = 1.07  # heat capacity of adsorbent [kJ/kgK]
        self.densa = 720  # density of adsorbent [kg/m3]
        self.Tadsinitial = 27  # initial temperature of adsorbent [C]
        self.mads = 54400  # total mass of the adsorbent in the column [kg]
        self.dma = self.mads / self.ny  # kg adsorbent per height segment

        # System conditions
        self.P = 65 * 10**2  # System pressure [kPa]
        self.Ppa = self.P * 1000  # System pressure [Pa]
        self.R = 8.314  # gas constant

        # Constants for Langmuir equation
        self.CH4C = np.array([
            [348.971, 0.542, 6.770E-06, 13672.210],
            [348.971, 0.542, 6.130E-07, 20307.220]
        ])
        self.H2OC = np.array([
            [-3799.940, 18.711, 3.580E-07, 44140.040],
            [3684.491, -4.450, 1.620E-05, 45199.990]
        ])

        # Gas properties
        self.Cpg = 2.73  # heat capacity of the gas [kJ/kgK]
        self.CpgSI = 2730  # Cpg in SI units [J/kgK]
        self.MWgas = 18.54  # MW of the gas
        self.MWwater = 18.02  # MW of water

        # Water properties
        self.CpH2O = 4.3  # heat capacity of water [kJ/kgK]

        # Ceramic ball properties
        self.mCpCB = 5500 * 0.84 / 13  # mCp ceramic balls in one height segment

        self.initialize_arrays()

    def initialize_arrays(self):
        self.mwg = np.zeros(self.ny)
        self.mwg1 = np.zeros(self.ny)
        self.Tgas = 27 * np.ones(self.ny)
        self.Hgas = np.zeros(self.ny)
        self.Hgas1 = np.zeros(self.ny)
        self.Tad = 27 * np.ones(self.ny)
        self.Mgas = np.zeros(self.ny)
        self.Mgas1 = np.zeros(self.ny)
        self.ppm = 0.1 * np.ones(self.ny)
        self.mwa = np.zeros(self.ny)

    def hfgas(self, x):
        return 4e-6 * x**3 - 0.0007 * x**2 + 2.6643 * x - 66.2325

    def densgasf(self, t):
        return np.full_like(t, 58.94)

    def viscgasf(self, u):
        return 1.326E-05

    def condgasf(self, v):
        return 3.840E-02

    def Tg(self, x):
        return -2.53123e-13 * x**5 + 4e-10 * x**4 - 3.33e-7 * x**3 + 9e-5 * x**2 + 0.3745 * x + 25

    def mwadp(self, YH2O, T):
        C11, C12, C13, C14 = self.CH4C[0]
        C21, C22, C23, C24 = self.CH4C[1]
        H11, H12, H13, H14 = self.H2OC[0]
        H21, H22, H23, H24 = self.H2OC[1]

        return 0.725 * (
            (H11 / (T + 273) + H12) * (H13 * np.exp(H14 / (self.R * (T + 273)))) * self.P * YH2O /
            (1 + ((H13 * np.exp(H14 / (self.R * (T + 273)))) * self.P * YH2O + (C13 * np.exp(C14 / (self.R * (T + 273)))) * self.P * (1 - YH2O))) +
            (H21 / (T + 273) + H22) * (H23 * np.exp(H24 / (self.R * (T + 273)))) * self.P * YH2O /
            (1 + ((H23 * np.exp(H24 / (self.R * (T + 273)))) * self.P * YH2O + (C23 * np.exp(C24 / (self.R * (T + 273)))) * self.P * (1 - YH2O)))
        ) * 18.015 / 1000

    def update_bed_saturation(self, t):
        dmwater = self.mwadp(730e-6, 27)
        MTZ_length = 0.98  # Length of MTZ in [m]

        if t <= 60 * 50:
            self.mwa[int(5*self.ny/6 - 2):] = self.dma * dmwater
            sigmoid_range = range(int(5*self.ny/6 - (MTZ_length*self.ny/self.h)), int(5*self.ny/6))
        elif t <= 60 * 100:
            self.mwa[int(4*self.ny/5 - 2):] = self.dma * dmwater
            sigmoid_range = range(int(4*self.ny/5 - (MTZ_length*self.ny/self.h)), int(4*self.ny/5))
        elif t <= 60 * 150:
            self.mwa[int(3*self.ny/4 - 2):] = self.dma * dmwater
            sigmoid_range = range(int(3*self.ny/4 - (MTZ_length*self.ny/self.h)), int(3*self.ny/4))
        elif t <= 60 * 200:
            self.mwa[int(2*self.ny/3 - 2):] = self.dma * dmwater
            sigmoid_range = range(int(2*self.ny/3 - (MTZ_length*self.ny/self.h)), int(2*self.ny/3))
        else:  # t <= 60 * 240
            self.mwa[int(self.ny/2 - 2):] = self.dma * dmwater
            sigmoid_range = range(int(self.ny/2 - (MTZ_length*self.ny/self.h)), int(self.ny/2))

        for i in sigmoid_range:
            sigmoid = (10 / len(sigmoid_range)) * (i - sigmoid_range.start) - 5
            self.mwa[i] = self.dma * dmwater * (1 / (1 + np.exp(-sigmoid)))

    def heat_mass_transfer(self):
        for j in range(self.ny):
            self.ppm[j] = np.clip((1e6 * (self.mwg[j] * self.MWgas / (self.Mgas[j] * self.MWwater))) if self.Mgas[j] > 0 else 0.1, 0.1, None)
            
            Tf = 0.5 * (self.Tgas[j] + self.Tad[j])
            q1 = 0.001 * self.dt * self.dy * 1402.28 * self.Qm * self.CpgSI * ((self.CpgSI * self.viscgasf(Tf) / self.condgasf(Tf)) ** (-2/3)) * \
                 ((2.53e-4 * self.Qm / self.viscgasf(Tf)) ** (-0.4069)) * (self.Tgas[j] - self.Tad[j])
            qmax = 0.92 * (self.Tgas[j] - self.Tad[j]) * (self.Cpg * self.Mgas[j] + self.CpH2O * self.mwg[j])
            q = np.clip(q1, qmax, q1) if q1 < 0 else np.clip(q1, None, qmax)

            if abs(self.mwg[j]) < 1e-12:
                self.mwg[j] = 0

            # Update temperatures
            epsilon = 1e-10  # Small value to prevent division by zero
            self.Tgas[j] -= q / (self.Cpg * self.Mgas[j] + self.CpH2O * self.mwg[j] + epsilon)
            self.Tad[j] += q / (self.Cpa * self.dma + self.CpH2O * self.mwa[j])

            # Mass transfer (simplified)
            dmwa = self.mwadp(self.ppm[j] * 1e-6, self.Tad[j]) - self.mwa[j] / self.dma
            transfer = 0.1 * dmwa * self.dma  # Assuming a transfer coefficient of 0.1
            self.mwa[j] += transfer
            self.mwg[j] -= transfer

    def gas_movement(self):
        densgasinitial = self.densgasf(self.Tadsinitial)
        vg = np.full(self.ny, self.Qm / (self.fA * densgasinitial))
        movecells = vg * self.dt / self.dy
        densgas = self.densgasf(self.Tgas)
        
        movegascells = movecells * self.Mgas
        moveHcells = movecells * self.Hgas
        movemwg = movecells * self.mwg
        
        for i in range(1, self.ny):
            self.Mgas1[i] = self.Mgas[i] - movegascells[i] + movegascells[i-1]
            self.Hgas1[i] = self.Hgas[i] - moveHcells[i] + moveHcells[i-1]
            self.mwg1[i] = self.mwg[i] - movemwg[i] + movemwg[i-1]

        Mgasentering = (1 - self.channeling) * self.Qm * self.dt
        self.Mgas1[0] = self.Mgas[0] - movegascells[0] + Mgasentering
        self.Hgas1[0] = self.Hgas[0] - moveHcells[0] + self.hfgas(27) * Mgasentering
        self.mwg1[0] = self.mwg[0] - movemwg[0]

        self.Mgas = self.Mgas1.copy()
        self.Hgas = self.Hgas1.copy()
        self.mwg = self.mwg1.copy()

    def run_simulation(self):
        time_array = np.zeros(self.nt)
        outlet_ppm_array = np.zeros(self.nt)
        bed_saturation_array = np.zeros((self.nt, self.ny))
        temperature_profile_array = np.zeros((self.nt, self.ny))

        for timestep in tqdm(range(1, self.nt + 1), desc="Simulation Progress"):
            t = self.dt * timestep
            
            self.update_bed_saturation(t)
            self.heat_mass_transfer()
            self.gas_movement()

            # Store data for plotting
            time_array[timestep-1] = t / 60  # Time in minutes
            outlet_ppm_array[timestep-1] = self.ppm[-1]
            bed_saturation_array[timestep-1] = self.mwa / (self.dma * self.mwadp(730e-6, 27)) * 100  # Percentage saturation
            temperature_profile_array[timestep-1] = self.Tgas

        return time_array, outlet_ppm_array, bed_saturation_array, temperature_profile_array

    def run_simulation(self):
        time_array = np.zeros(self.nt)
        outlet_ppm_array = np.zeros(self.nt)
        bed_saturation_array = np.zeros((self.nt, self.ny))
        temperature_profile_array = np.zeros((self.nt, self.ny))

        for timestep in tqdm(range(1, self.nt + 1), desc="Simulation Progress"):
            t = self.dt * timestep
            
            self.update_bed_saturation(t)
            self.heat_mass_transfer()
            self.gas_movement()

            # Store data for plotting
            time_array[timestep-1] = t / 60  # Time in minutes
            outlet_ppm_array[timestep-1] = self.ppm[-1]
            bed_saturation_array[timestep-1] = self.mwa / (self.dma * self.mwadp(730e-6, 27)) * 100  # Percentage saturation
            temperature_profile_array[timestep-1] = self.Tgas

        return time_array, outlet_ppm_array, bed_saturation_array, temperature_profile_array

    def plot_results(self, time_array, outlet_ppm_array, bed_saturation_array, temperature_profile_array):
        plt.figure(figsize=(15, 10))

        # Plot 1: Bed Saturation Profile
        plt.subplot(2, 2, 1)
        aaa = np.linspace(0, self.h, self.ny)
        plt.plot(aaa, bed_saturation_array[-1], 'g')
        plt.title(f'Adsorbent Bed Saturation at t={self.tmax/60:.0f}mins, q={self.Qmh}kg/h')
        plt.xlabel('Position in column [m]')
        plt.ylabel('Adsorbent Bed Saturation [%]')
        plt.grid(True)

        # Plot 2: Temperature Profile
        plt.subplot(2, 2, 2)
        plt.plot(aaa, temperature_profile_array[-1], 'r')
        plt.title(f'Temperature Profile at t={self.tmax/60:.0f}mins')
        plt.xlabel('Position in column [m]')
        plt.ylabel('Temperature [C]')
        plt.grid(True)

        # Plot 3: Outlet PPM over time
        plt.subplot(2, 2, 3)
        plt.plot(time_array, outlet_ppm_array, 'b')
        plt.title('Outlet PPM over Time')
        plt.xlabel('Time [minutes]')
        plt.ylabel('Outlet PPM')
        plt.grid(True)

        # Plot 4: Bed Saturation over time (heatmap)
        plt.subplot(2, 2, 4)
        plt.imshow(bed_saturation_array.T, aspect='auto', origin='lower', 
                   extent=[0, self.tmax/60, 0, self.h], cmap='viridis')
        plt.colorbar(label='Bed Saturation [%]')
        plt.title('Bed Saturation over Time')
        plt.xlabel('Time [minutes]')
        plt.ylabel('Position in column [m]')

        plt.tight_layout()
        plt.show()

if __name__ == "__main__":
    # Create simulation instance
    simulation = AdsorptionColumnSimulation()

    # Run simulation
    time_array, outlet_ppm_array, bed_saturation_array, temperature_profile_array = simulation.run_simulation()

    # Plot results
    simulation.plot_results(time_array, outlet_ppm_array, bed_saturation_array, temperature_profile_array)
```
