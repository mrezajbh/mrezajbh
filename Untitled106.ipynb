```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize, differential_evolution
from scipy.optimize import fsolve
from sympy import symbols, diff, exp, log, solve
import sympy as sp

# =============================================================================
# Constants and Data
# =============================================================================

# Temperature of the system
T = 353.15  # K

# Experimental data
x1_exp = np.array([0, 0.0339, 0.0653, 0.1538, 0.2408, 0.2994, 0.3483, 0.3958,
                  0.4511, 0.5171, 0.5979, 0.6497, 0.7089, 0.7767, 0.9217, 1])
P_exp = 1e-3 * np.array([84, 109, 135, 208, 276, 322, 360, 411, 465, 525, 590,
                         636, 683, 736, 845, 894])  # bar
y1_exp = np.array([0, 0.2486, 0.4203, 0.6480, 0.7576, 0.8036, 0.8483, 0.8660,
                  0.8921, 0.9171, 0.9387, 0.9462, 0.9589, 0.9748, 0.9914, 1])

# Antoine equation constants
Aa = np.array([7.1557, 7.5253])
B = np.array([1274.31, 1693.50])
C = np.array([-50.040, -50.661])

# Virial coefficients
B11 = -922.5
B22 = -2104.9
B12 = -1285.1
del12 = (2 * B12) - B11 - B22

# Other parameters
Rg = 83.14  # Gas constant, cm³·bar/(mol·K)
V = np.array([77.59, 112.03])  # Molar volumes, cm³/mol

# Pure component vapor pressures
Prs = 1e-3 * 10**(Aa - (B / (T + C)))  # bar

# Fugacity coefficients
k1 = B11 / (Rg * T)
k2 = B22 / (Rg * T)
fis = np.exp([k1 * Prs[0], k2 * Prs[1]])

# Activity coefficient experimental data for NRTL and Wilson models (excluding first and last points)
gamaexp1 = np.array([0.916028964702250, 0.995227124415963, 1.00243796800111,
                     0.992249371970922, 0.986914486099804, 1.00083207597047,
                     1.02561241721250, 1.04803150769957, 1.06034435253726,
                     1.05399477935872, 1.05322852607261, 1.04992786332553,
                     1.04917962259582, 1.03092124370248])

gamaexp2 = np.array([1.00271851067249, 0.988210267986742, 1.01557535913408,
                     1.02915998043412, 1.05067097767073, 0.972677301302471,
                     1.05414493164453, 1.05298216333068, 1.03375155281564,
                     1.02684475339959, 1.11145243869106, 1.09356339635854,
                     0.938336239104532, 1.04031729938681])

# Composition data for NRTL and Wilson models (excluding first and last points)
xcomp = np.array([0.0339, 0.0653, 0.1538, 0.2408, 0.2994, 0.3483, 0.3958, 0.4511,
                  0.5171, 0.5979, 0.6497, 0.7089, 0.7767, 0.9217])

# =============================================================================
# NRTL Model
# =============================================================================

def optnrtl1(x):
    """
    NRTL model optimization function for binary system

    Parameters:
    x (array): Model parameters [tau12, tau21, alpha]

    Returns:
    float: Sum of absolute errors between experimental and calculated activity coefficients
    """
    # Initialize arrays
    gamanrtl = np.zeros((2, len(xcomp)))
    ernrtl = np.zeros((2, len(xcomp)))

    # Calculate activity coefficients using NRTL model
    for j in range(len(xcomp)):
        # Component 1
        gamanrtl[0, j] = np.exp(
            ((1-xcomp[j])**2) * (
                x[1] * (((np.exp(-x[2]*x[1])) / (xcomp[j] + ((np.exp(-x[2]*x[1])) * (1-xcomp[j]))))**2)
            ) + (
                (x[0] * np.exp(-x[2]*x[0])) / ((1-xcomp[j]) + (np.exp(-x[2]*x[0]) * xcomp[j]))**2
            )
        )

        # Component 2
        gamanrtl[1, j] = np.exp(
            ((xcomp[j])**2) * (
                x[0] * (((np.exp(-x[2]*x[0])) / ((1-xcomp[j]) + ((np.exp(-x[2]*x[0])) * xcomp[j])))**2)
            ) + (
                (x[1] * np.exp(-x[2]*x[1])) / (xcomp[j] + (np.exp(-x[2]*x[1]) * (1-xcomp[j])))**2
            )
        )

        # Calculate errors
        ernrtl[0, j] = abs(gamanrtl[0, j] - gamaexp1[j])
        ernrtl[1, j] = abs(gamanrtl[1, j] - gamaexp2[j])

    # Return sum of errors for component 1
    return np.sum(ernrtl[0, :])

def calculate_nrtl_activity_coefficients(x1, params):
    """
    Calculate activity coefficients using the NRTL model for a binary system

    Parameters:
    x1 (array): Mole fractions of component 1
    params (array): NRTL parameters [tau12, tau21, alpha]

    Returns:
    tuple: Activity coefficients for components 1 and 2
    """
    x2 = 1 - x1
    tau12, tau21, alpha = params

    gamma1 = np.zeros_like(x1)
    gamma2 = np.zeros_like(x1)

    for i in range(len(x1)):
        G12 = np.exp(-alpha * tau12)
        G21 = np.exp(-alpha * tau21)

        # Activity coefficient for component 1
        term1 = x2[i]**2 * tau21 * G21 / (x1[i] + x2[i] * G21)**2
        term2 = tau12 * G12 / (G12 * x1[i] + x2[i])**2
        gamma1[i] = np.exp(x2[i]**2 * (term1 + term2))

        # Activity coefficient for component 2
        term1 = x1[i]**2 * tau12 * G12 / (x2[i] + x1[i] * G12)**2
        term2 = tau21 * G21 / (G21 * x2[i] + x1[i])**2
        gamma2[i] = np.exp(x1[i]**2 * (term1 + term2))

    return gamma1, gamma2

# =============================================================================
# Wilson Model
# =============================================================================

def optwilson1(x):
    # Initialize arrays correctly
    gamawilson = np.zeros((2, len(xcomp)))
    finalerorwilson = np.zeros(len(xcomp))

    for j in range(len(xcomp)):
        # Component 1 - exactly matching MATLAB formula
        gamawilson[0, j] = np.exp(
            -np.log(1 - (x[1] * (1 - xcomp[j]))) +
            ((1 - xcomp[j]) * (
                (x[0] * (1 - xcomp[j]) / (1 - (x[0] * xcomp[j]))) -
                (x[1] * xcomp[j] / (1 - (x[1] * (1 - xcomp[j]))))
            ))
        )

        # Component 2
        gamawilson[1, j] = np.exp(
            -np.log(1 - (x[0] * xcomp[j])) -
            (xcomp[j] * (
                (x[0] * (1 - xcomp[j]) / (1 - (x[0] * xcomp[j]))) -
                (x[1] * xcomp[j] / (1 - (x[1] * (1 - xcomp[j]))))
            ))
        )

        finalerorwilson[j] = abs(gamawilson[0, j] - gamaexp1[j])

    # Return sum of errors for component 1
    return np.sum(finalerorwilson)



def calculate_wilson_activity_coefficients(x1, params):
    """
    Calculate activity coefficients using the Wilson model for a binary system

    Parameters:
    x1 (array): Mole fractions of component 1
    params (array): Wilson parameters [Lambda12, Lambda21]

    Returns:
    tuple: Activity coefficients for components 1 and 2
    """
    x2 = 1 - x1
    Lambda12, Lambda21 = params

    gamma1 = np.zeros_like(x1)
    gamma2 = np.zeros_like(x1)

    for i in range(len(x1)):
        # Component 1 - exactly matching MATLAB formula
        gamma1[i] = np.exp(
            -np.log(1 - (Lambda21 * x2[i])) +
            (x2[i] * (
                (Lambda12 * x2[i] / (1 - (Lambda12 * x1[i]))) -
                (Lambda21 * x1[i] / (1 - (Lambda21 * x2[i])))
            ))
        )

        # Component 2
        gamma2[i] = np.exp(
            -np.log(1 - (Lambda12 * x1[i])) -
            (x1[i] * (
                (Lambda12 * x2[i] / (1 - (Lambda12 * x1[i]))) -
                (Lambda21 * x1[i] / (1 - (Lambda21 * x2[i])))
            ))
        )

    return gamma1, gamma2

# =============================================================================
# UNIFAC Model
# =============================================================================

def optunifac1(x):
    """
    UNIFAC model optimization function

    Parameters:
    x (array): Group interaction parameters (16 parameters in a 4x4 matrix)

    Returns:
    float: Sum of absolute errors between experimental and calculated activity coefficients
    """
    # Reshape parameters into 4x4 matrix
    a = x.reshape(4, 4)

    # UNIFAC Model parameters
    R = np.array([0.5313, 0.3652, 1.1450, 1.4654])
    Q = np.array([0.4, 0.120, 1.088, 1.264])

    # Component group assignments (m_ik: number of groups k in molecule i)
    m = np.array([
        [5, 1, 1, 0],  # Component 1
        [0, 0, 0, 2]   # Component 2
    ])

    # Calculate r_i and q_i (van der Waals volume and surface area parameters)
    r = np.zeros(2)
    q = np.zeros(2)
    for i in range(2):
        r[i] = np.sum(m[i, :] * R)
        q[i] = np.sum(m[i, :] * Q)

    # Experimental data for all compositions
    xcomp_full = np.array([
        [0, 0.0339, 0.0653, 0.1538, 0.2408, 0.2994, 0.3483, 0.3958, 0.4511, 0.5171, 0.5979, 0.6497, 0.7089, 0.7767, 0.9217, 1],
        [1, 0.9661, 0.9347, 0.8462, 0.7592, 0.7006, 0.6517, 0.6042, 0.5489, 0.4829, 0.4021, 0.3503, 0.2911, 0.2233, 0.0783, 0]
    ])

    # Experimental activity coefficients (excluding first and last points)
    gamaexpermental = np.array([
        [float('nan'), 0.916028964702250, 0.995227124415963, 1.00243796800111, 0.992249371970922,
         0.986914486099804, 1.00083207597047, 1.02561241721250, 1.04803150769957, 1.06034435253726,
         1.05399477935872, 1.05322852607261, 1.04992786332553, 1.04917962259582, 1.03092124370248, 1.01346229203214],
        [0.995664361461433, 1.00271851067249, 0.988210267986742, 1.01557535913408, 1.02915998043412,
         1.05067097767073, 0.972677301302471, 1.05414493164453, 1.05298216333068, 1.03375155281564,
         1.02684475339959, 1.11145243869106, 1.09356339635854, 0.938336239104532, 1.04031729938681, float('nan')]
    ])

    # Initialize arrays for results
    gamacom = np.zeros((2, 16))

    # Calculate combinatorial part of activity coefficients
    for j in range(16):
        s = np.zeros(2)  # Total molecular volume
        t = np.zeros(2)  # Total molecular surface area

        for i in range(2):
            s[i] = np.sum(r[i] * xcomp_full[i, j])
            t[i] = np.sum(q[i] * xcomp_full[i, j])

        fi = np.zeros(2)   # Molecular volume fraction
        teta = np.zeros(2) # Molecular surface area fraction
        l = np.zeros(2)

        for i in range(2):
            fi[i] = xcomp_full[i, j] * r[i] / s[i]
            teta[i] = xcomp_full[i, j] * q[i] / t[i]
            l[i] = 5 * (r[i] - q[i]) - (r[i] - 1)

        y = np.sum(l * xcomp_full[:, j])

        for i in range(2):
            gamacom[i, j] = np.exp(
                np.log(fi[i] / xcomp_full[i, j]) +
                5 * q[i] * np.log(teta[i] / fi[i]) +
                l[i] - (fi[i] / xcomp_full[i, j]) * y
            )

    # Calculate residual part of activity coefficients
    si = np.exp(-a / T)  # Segment fraction

    # Create array for group fractions
    p = np.zeros((8, 16))
    for j in range(16):
        for i in range(4):
            p[i, j] = m[0, i] * xcomp_full[0, j]
            p[i+4, j] = m[1, i] * xcomp_full[1, j]

    # Calculate liquid phase group fractions and other parameters
    X = np.zeros((4, 16))
    u = np.zeros((4, 16))
    TETA = np.zeros((4, 16))

    for j in range(16):
        P_sum = np.sum(p[:, j])
        for i in range(4):
            X[i, j] = (p[i, j] + p[i+4, j]) / P_sum
            u[i, j] = Q[i] * X[i, j]

        U = np.sum(u[:, j])
        for i in range(4):
            TETA[i, j] = Q[i] * X[i, j] / U

    # Calculate group interaction terms
    d = np.zeros((4, 16, 4))
    c = np.zeros((4, 16, 4))

    for j in range(16):
        for i in range(4):
            for k in range(4):
                d[k, j, i] = TETA[k, j] * si[k, i]
                c[k, j, i] = TETA[k, j] * si[i, k]

    # Calculate group activity coefficients
    GAMA = np.zeros((4, 16))
    for j in range(16):
        for i in range(4):
            d_sum = np.sum(d[:, j, i])
            c_sum = np.sum(c[:, j, i])
            w = c[:, j, i] / c_sum
            GAMA[i, j] = np.exp(Q[i] * (1 - np.log(d_sum) - np.sum(w)))

    # Calculate pure component group activity coefficients
    Xpure = np.zeros((2, 4))
    TETApure = np.zeros((2, 4))

    for i in range(2):
        for j in range(4):
            Xpure[i, j] = m[i, j] / np.sum(m[i, :])
            TETApure[i, j] = Q[j] * Xpure[i, j] / np.sum(Q * Xpure[i, :])

    # Calculate group interaction terms for pure components
    o1 = np.zeros((4, 4))
    o2 = np.zeros((4, 4))
    n1 = np.zeros((4, 4))
    n2 = np.zeros((4, 4))

    for i in range(4):
        for k in range(4):
            o1[i, k] = TETApure[0, k] * si[k, i]
            o2[i, k] = TETApure[1, k] * si[k, i]
            n1[i, k] = TETApure[0, k] * si[i, k]
            n2[i, k] = TETApure[1, k] * si[i, k]

    # Calculate pure component group interaction ratios
    e1 = np.zeros((4, 4))
    e2 = np.zeros((4, 4))

    for k in range(4):
        for i in range(4):
            e1[k, i] = n1[k, i] / np.sum(o1[i, :])
            e2[k, i] = n2[k, i] / np.sum(o2[i, :])

    # Calculate pure component group activity coefficients
    gamarespure = np.zeros((2, 4))

    for i in range(4):
        gamarespure[0, i] = Q[i] * (1 - np.log(np.sum(o1[i, :])) - np.sum(e1[i, :]))
        gamarespure[1, i] = Q[i] * (1 - np.log(np.sum(o2[i, :])) - np.sum(e2[i, :]))

    # Calculate residual and total activity coefficients
    gamares = np.zeros((2, 16))
    gama = np.zeros((2, 16))

    for j in range(16):
        for i in range(2):
            gamares_sum = 0
            for k in range(4):
                gamares_sum += m[i, k] * (GAMA[k, j] - gamarespure[i, k])
            gamares[i, j] = np.exp(gamares_sum)
            gama[i, j] = np.exp(np.log(gamacom[i, j]) + np.log(gamares[i, j]))

    # Calculate error (excluding NaN values)
    finaleror = np.zeros(16)
    for j in range(16):
        if not np.isnan(gamaexpermental[0, j]):
            finaleror[j] = abs(gama[0, j] - gamaexpermental[0, j])

    return np.nansum(finaleror)

# =============================================================================
# Excess Gibbs Energy Modeling
# =============================================================================

def opt1(x):
    """
    Optimization function for excess Gibbs energy model

    Parameters:
    x (array): Legendre polynomial expansion coefficients

    Returns:
    float: Sum of relative errors between experimental and calculated pressures
    """
    # Define Legendre polynomials
    L0 = np.ones_like(x1_exp)
    L1 = (2 * x1_exp) - 1
    L2 = 6 * (x1_exp**2 - x1_exp + (1/6))
    L3 = 20 * (x1_exp**3 - ((3/2) * x1_exp**2) + ((3/5) * x1_exp) - (1/20))
    L4 = 70 * (x1_exp**4 - (2 * x1_exp**3) + ((9/7) * x1_exp**2) + (1/70))

    # Calculate excess Gibbs energy
    M = x[0] * L0 + x[1] * L1 + x[2] * L2 + x[3] * L3 + x[4] * L4
    molgibs = x1_exp * (1 - x1_exp) * M

    # Calculate derivative of excess Gibbs energy
    molgibsprime = np.zeros_like(x1_exp)
    for j in range(len(x1_exp)):
        molgibsprime[j] = ((1 - (2 * x1_exp[j])) * M[j]) + \
                          ((x1_exp[j] - x1_exp[j]**2) * (
                              (2 * x[1]) +
                              (12 * x[2] * x1_exp[j]) - (6 * x[2]) +
                              (60 * x[3] * x1_exp[j]**2) - (60 * x[3] * x1_exp[j]) + (12 * x[3]) +
                              (280 * x[4] * x1_exp[j]**3) - (420 * x[4] * x1_exp[j]**2) + (180 * x[4] * x1_exp[j]) - (20 * x[4])
                          ))

    # Calculate vapor phase fugacity coefficients
    x2_exp = 1 - x1_exp
    y2_exp = 1 - y1_exp

    fiv1 = np.zeros_like(x1_exp)
    fiv2 = np.zeros_like(x1_exp)

    for j in range(len(x1_exp)):
        fiv1[j] = np.exp((P_exp[j] / (Rg * T)) * (B11 + ((y2_exp[j]**2) * del12)))
        fiv2[j] = np.exp((P_exp[j] / (Rg * T)) * (B22 + ((y1_exp[j]**2) * del12)))

    # Calculate pressures
    Pressure = np.zeros_like(x1_exp)
    for j in range(len(x1_exp)):
        Pressure[j] = (Prs[0] * (fis[0] / fiv1[j]) *
                       np.exp(molgibs[j] + (molgibsprime[j] * x2_exp[j]) +
                             (V[0] * (P_exp[j] - Prs[0]) / (Rg * T))) *
                       x1_exp[j]) + \
                      (Prs[1] * (fis[1] / fiv2[j]) *
                       np.exp(molgibs[j] - (molgibsprime[j] * x1_exp[j]) +
                             (V[1] * (P_exp[j] - Prs[1]) / (Rg * T))) *
                       x2_exp[j])

    # Calculate relative errors
    error = np.abs(Pressure - P_exp) / P_exp

    return np.sum(error)

def legendr_derivative(x1):
    """
    Calculate derivative of the Legendre polynomial expansion for excess Gibbs energy

    Parameters:
    x1 (float or array): Mole fraction of component 1

    Returns:
    float or array: Derivative of excess Gibbs energy function
    """
    # Define molar Gibbs energy function using the coefficients [0.0064, 0.0048, 0, 0.0018, 0.0002]
    x = np.array([0.0064, 0.0048, 0, 0.0018, 0.0002])

    # Analytical derivative
    molgibsenergyprime = (
        -((7 * x1**4) / 500 + x1**3 / 125 - (9 * x1**2) / 250 + (39 * x1) / 1250) +
        (x1 - 1) * ((7 * x1**3) / 125 + (3 * x1**2) / 125 - (9 * x1) / 125 + 39 / 1250) -
        x1 * ((7 * x1**3) / 125 + (3 * x1**2) / 125 - (9 * x1) / 125 + 39 / 1250)
    )

    return molgibsenergyprime

# =============================================================================
# Optimization and Calculation Functions
# =============================================================================

def optimize_nrtl_parameters():
    """
    Optimize NRTL model parameters

    Returns:
    array: Optimized NRTL parameters [tau12, tau21, alpha]
    """
    # Initial guess
    x0 = [0.5, 0.5, 0.3]

    # Parameter bounds
    bounds = [(-1, 5), (-1, 5), (0, 1)]

    # Run differential evolution
    result = differential_evolution(optnrtl1, bounds, strategy='best1bin',
                                   maxiter=1000, popsize=15, tol=0.01)

    return result.x

def optimize_wilson_parameters():

    bounds = [(-10, 1), (-10, 1)]

    result = differential_evolution(optwilson1, bounds, strategy='best1bin',
                                   maxiter=1000, popsize=15, tol=0.01)
    return result.x

def optimize_unifac_parameters():
    """
    Optimize UNIFAC model parameters

    Returns:
    array: Optimized UNIFAC parameters (16 values for 4x4 interaction matrix)
    """
    # Initial guess - all zeros
    x0 = np.zeros(16)

    # Parameter bounds
    bounds = [(-3000, 4000)] * 16

    # Run differential evolution
    result = differential_evolution(optunifac1, bounds, strategy='best1bin',
                                   maxiter=300, popsize=20, tol=0.1)

    return result.x

def optimize_gibbs_parameters():
    """
    Optimize excess Gibbs energy model parameters

    Returns:
    array: Optimized Gibbs energy parameters (5 coefficients)
    """
    # Initial guess
    x0 = [0.01, 0.01, 0.01, 0.01, 0.01]

    # Parameter bounds
    bounds = [(0, 1000)] * 5

    # Run differential evolution
    result = differential_evolution(opt1, bounds, strategy='best1bin',
                                   maxiter=500, popsize=15, tol=0.01)

    return result.x

def calculate_vle_using_nrtl(nrtl_params, T=353.15, x1_range=None):
    """
    Calculate VLE data using the NRTL model

    Parameters:
    nrtl_params (array): NRTL parameters [tau12, tau21, alpha]
    T (float): Temperature in K
    x1_range (array, optional): Array of liquid phase mole fractions

    Returns:
    tuple: (x1, y1, P) arrays of liquid compositions, vapor compositions, and pressures
    """
    if x1_range is None:
        x1_range = np.linspace(0.01, 0.99, 50)

    # Initialize arrays
    y1 = np.zeros_like(x1_range)
    P = np.zeros_like(x1_range)

    # Calculate activity coefficients
    gamma1, gamma2 = calculate_nrtl_activity_coefficients(x1_range, nrtl_params)

    # Calculate VLE using modified Raoult's law with activity coefficients
    for i in range(len(x1_range)):
        # Calculate pressure
        P[i] = x1_range[i] * gamma1[i] * Prs[0] + (1 - x1_range[i]) * gamma2[i] * Prs[1]

        # Calculate vapor composition
        y1[i] = x1_range[i] * gamma1[i] * Prs[0] / P[i]

        # Adjust for nonideality in vapor phase using virial equation
        # This requires an iterative solution, simplified here
        error = 1
        y1_old = y1[i]
        while error > 1e-4:
            # Calculate fugacity coefficients
            fiv1_val = np.exp((P[i] / (Rg * T)) * (B11 + ((1 - y1_old)**2 * del12)))
            fiv2_val = np.exp((P[i] / (Rg * T)) * (B22 + (y1_old**2 * del12)))

            # Recalculate vapor composition
            y1_new = x1_range[i] * gamma1[i] * Prs[0] / (P[i] * fiv1_val)

            # Check convergence
            error = abs(y1_new - y1_old)
            y1_old = y1_new

            # Break if not converging
            if error == 1:
                break

        y1[i] = y1_old

    return x1_range, y1, P

def calculate_vle_using_wilson(wilson_params, T=353.15, x1_range=None):
    """
    Calculate VLE data using the Wilson model

    Parameters:
    wilson_params (array): Wilson parameters [Lambda12, Lambda21]
    T (float): Temperature in K
    x1_range (array, optional): Array of liquid phase mole fractions

    Returns:
    tuple: (x1, y1, P) arrays of liquid compositions, vapor compositions, and pressures
    """
    if x1_range is None:
        x1_range = np.linspace(0.01, 0.99, 50)

    # Initialize arrays
    y1 = np.zeros_like(x1_range)
    P = np.zeros_like(x1_range)

    # Calculate activity coefficients
    gamma1, gamma2 = calculate_wilson_activity_coefficients(x1_range, wilson_params)

    # Calculate VLE using modified Raoult's law with activity coefficients
    for i in range(len(x1_range)):
        # Calculate pressure
        P[i] = x1_range[i] * gamma1[i] * Prs[0] + (1 - x1_range[i]) * gamma2[i] * Prs[1]

        # Calculate vapor composition
        y1[i] = x1_range[i] * gamma1[i] * Prs[0] / P[i]

        # Adjust for nonideality in vapor phase using virial equation
        # This requires an iterative solution, simplified here
        error = 1
        y1_old = y1[i]
        while error > 1e-4:
            # Calculate fugacity coefficients
            fiv1_val = np.exp((P[i] / (Rg * T)) * (B11 + ((1 - y1_old)**2 * del12)))
            fiv2_val = np.exp((P[i] / (Rg * T)) * (B22 + (y1_old**2 * del12)))

            # Recalculate vapor composition
            y1_new = x1_range[i] * gamma1[i] * Prs[0] / (P[i] * fiv1_val)

            # Check convergence
            error = abs(y1_new - y1_old)
            y1_old = y1_new

            # Break if not converging
            if error == 1:
                break

        y1[i] = y1_old

    return x1_range, y1, P

def calculate_activity_coefficients_from_experimental_data():
    """
    Calculate experimental activity coefficients from VLE data

    Returns:
    tuple: (gamma1, gamma2) arrays of activity coefficients for components 1 and 2
    """
    # Calculate vapor phase fugacity coefficients
    fiv1 = np.zeros_like(x1_exp)
    fiv2 = np.zeros_like(x1_exp)

    for j in range(len(x1_exp)):
        fiv1[j] = np.exp((P_exp[j] / (Rg * T)) * (B11 + ((1 - y1_exp[j])**2 * del12)))
        fiv2[j] = np.exp((P_exp[j] / (Rg * T)) * (B22 + (y1_exp[j]**2 * del12)))

    # Calculate activity coefficients
    gamma1 = np.zeros_like(x1_exp)
    gamma2 = np.zeros_like(x1_exp)

    for j in range(len(x1_exp)):
        if x1_exp[j] > 0:
            gamma1[j] = fiv1[j] * y1_exp[j] * P_exp[j] / (x1_exp[j] * Prs[0] * fis[0])
        if x1_exp[j] < 1:
            gamma2[j] = fiv2[j] * (1 - y1_exp[j]) * P_exp[j] / ((1 - x1_exp[j]) * Prs[1] * fis[1])

    return gamma1, gamma2

# =============================================================================
# Plotting Functions
# =============================================================================

def plot_pxy_diagram(x1_exp, y1_exp, P_exp, x1_calc=None, y1_calc=None, P_calc=None, title="P-x-y Diagram"):
    """
    Plot a P-x-y diagram

    Parameters:
    x1_exp (array): Experimental liquid mole fractions
    y1_exp (array): Experimental vapor mole fractions
    P_exp (array): Experimental pressures
    x1_calc (array, optional): Calculated liquid mole fractions
    y1_calc (array, optional): Calculated vapor mole fractions
    P_calc (array, optional): Calculated pressures
    title (str, optional): Plot title
    """
    plt.figure(figsize=(10, 6))

    # Plot experimental data
    plt.scatter(x1_exp, P_exp, marker='o', color='blue', label='Experimental (liquid)')
    plt.scatter(y1_exp, P_exp, marker='^', color='red', label='Experimental (vapor)')

    # Plot calculated data if provided
    if x1_calc is not None and P_calc is not None:
        plt.plot(x1_calc, P_calc, 'b-', label='Calculated (liquid)')
    if y1_calc is not None and P_calc is not None:
        plt.plot(y1_calc, P_calc, 'r-', label='Calculated (vapor)')

    plt.xlabel('Mole Fraction of Component 1')
    plt.ylabel('Pressure (bar)')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_xy_diagram(x1_exp, y1_exp, x1_calc=None, y1_calc=None, title="x-y Diagram"):
    """
    Plot an x-y diagram

    Parameters:
    x1_exp (array): Experimental liquid mole fractions
    y1_exp (array): Experimental vapor mole fractions
    x1_calc (array, optional): Calculated liquid mole fractions
    y1_calc (array, optional): Calculated vapor mole fractions
    title (str, optional): Plot title
    """
    plt.figure(figsize=(8, 8))

    # Plot experimental data
    plt.scatter(x1_exp, y1_exp, marker='o', color='blue', label='Experimental')

    # Plot calculated data if provided
    if x1_calc is not None and y1_calc is not None:
        plt.plot(x1_calc, y1_calc, 'r-', label='Calculated')

    # Plot diagonal line (y=x)
    plt.plot([0, 1], [0, 1], 'k--', label='y=x')

    plt.xlabel('Liquid Mole Fraction (x₁)')
    plt.ylabel('Vapor Mole Fraction (y₁)')
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_activity_coefficients(x1, gamma1, gamma2, model_name=""):
    """
    Plot activity coefficients vs composition

    Parameters:
    x1 (array): Liquid mole fractions
    gamma1 (array): Activity coefficients for component 1
    gamma2 (array): Activity coefficients for component 2
    model_name (str, optional): Name of the model for the title
    """
    plt.figure(figsize=(8, 6))

    plt.plot(x1, gamma1, 'b-', label='Component 1')
    plt.plot(x1, gamma2, 'r-', label='Component 2')

    plt.xlabel('Liquid Mole Fraction (x₁)')
    plt.ylabel('Activity Coefficient (γ)')
    if model_name:
        plt.title(f'Activity Coefficients vs Composition ({model_name} Model)')
    else:
        plt.title('Activity Coefficients vs Composition')
    plt.legend()
    plt.grid(True)
    plt.show()










# =============================================================================
# Main Execution
# =============================================================================

def main():
    """Main function to execute the VLE calculations and display results"""
    print("VLE Thermodynamic Modeling in Python")
    print("=" * 40)

    # 1. Calculate experimental activity coefficients
    gamma1_exp, gamma2_exp = calculate_activity_coefficients_from_experimental_data()
    print("\nExperimental Activity Coefficients (sample):")
    for i in range(0, len(x1_exp), 3):
        if i < len(x1_exp):
            print(f"x1={x1_exp[i]:.4f}: gamma1={gamma1_exp[i]:.4f}, gamma2={gamma2_exp[i]:.4f}")

    # 2. Optimize NRTL parameters
    print("\nOptimizing NRTL parameters...")
    nrtl_params = optimize_nrtl_parameters()
    print(f"Optimized NRTL parameters: tau12={nrtl_params[0]:.4f}, tau21={nrtl_params[1]:.4f}, alpha={nrtl_params[2]:.4f}")

    # 3. Optimize Wilson parameters
    print("\nOptimizing Wilson parameters...")
    wilson_params = optimize_wilson_parameters()
    print(f"Optimized Wilson parameters: Lambda12={wilson_params[0]:.4f}, Lambda21={wilson_params[1]:.4f}")

    # 4. Calculate VLE using NRTL model
    print("\nCalculating VLE using NRTL model...")
    x1_nrtl, y1_nrtl, P_nrtl = calculate_vle_using_nrtl(nrtl_params)

    # 5. Calculate VLE using Wilson model
    print("\nCalculating VLE using Wilson model...")
    x1_wilson, y1_wilson, P_wilson = calculate_vle_using_wilson(wilson_params)

    # 6. Plot P-x-y diagram with NRTL model
    plot_pxy_diagram(x1_exp, y1_exp, P_exp, x1_nrtl, y1_nrtl, P_nrtl, "P-x-y Diagram (NRTL Model)")

    # 7. Plot P-x-y diagram with Wilson model
    plot_pxy_diagram(x1_exp, y1_exp, P_exp, x1_wilson, y1_wilson, P_wilson, "P-x-y Diagram (Wilson Model)")

    # 8. Plot x-y diagram with NRTL model
    plot_xy_diagram(x1_exp, y1_exp, x1_nrtl, y1_nrtl, "x-y Diagram (NRTL Model)")

    # 9. Plot x-y diagram with Wilson model
    plot_xy_diagram(x1_exp, y1_exp, x1_wilson, y1_wilson, "x-y Diagram (Wilson Model)")

    # 10. Plot activity coefficients
    gamma1_nrtl, gamma2_nrtl = calculate_nrtl_activity_coefficients(x1_nrtl, nrtl_params)
    gamma1_wilson, gamma2_wilson = calculate_wilson_activity_coefficients(x1_wilson, wilson_params)

    plot_activity_coefficients(x1_nrtl, gamma1_nrtl, gamma2_nrtl, "NRTL")
    plot_activity_coefficients(x1_wilson, gamma1_wilson, gamma2_wilson, "Wilson")

    print("\nCalculations and plots complete.")

if __name__ == "__main__":
    main()







```
