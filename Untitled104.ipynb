```python
import pandas as pd
import numpy as np
import csv

# ============================================================================
# Choose the data source for the renewable production data (now using ninja)
# ============================================================================
data_source = "ninja"  # Using ninja datasets

# Update the file paths to the ninja datasets
solar_file = "ninja_pv_country_DK_merra-2_nuts-2_corrected.csv"
wind_file = "ninja_wind_country_DK_current_merra-2_nuts-2_corrected.csv"

# Read the CSV files, skipping the first two rows which contain metadata
# For PV data
solar_data = []
with open(solar_file, 'r') as f:
    lines = f.readlines()
    pv_headers = lines[2].strip().split(',')
    for line in lines[3:]:
        if line.strip() and line.startswith('2018'):  # Using 2018 data
            values = line.strip().split(',')
            solar_data.append(values)

# For wind data
wind_data = []
with open(wind_file, 'r') as f:
    lines = f.readlines()
    wind_headers = lines[2].strip().split(',')
    for line in lines[3:]:
        if line.strip() and line.startswith('2018'):  # Using 2018 data
            values = line.strip().split(',')
            wind_data.append(values)

# Extract the capacity factors - using DK01 for PV and DK_TOTAL for wind if available
pv_column_idx = pv_headers.index('DK01')
wind_column_idx = wind_headers.index('DK_TOTAL') if 'DK_TOTAL' in wind_headers else wind_headers.index('DK01')

# Extract hourly production data
solar_power_1kwp = [float(row[pv_column_idx]) for row in solar_data][:8760]
wind_power_1kwp = [float(row[wind_column_idx]) for row in wind_data][:8760]

# ============================================================================
# Simulation time settings
# ============================================================================
dt    = 1      # time step in hours
t_sim = 8760   # simulation period (1 year)

# TAD threshold (Total Ammonia Deficit) to limit iterations
# ADJUSTED: Increased to allow more hours of deficit since all configurations exceed the original threshold
TAD_threshold      = 24 * 300       # 300 days worth of hours (changed from 5 days)
TAD_threshold_days = TAD_threshold / 24

print(f"Using adjusted TAD threshold of {TAD_threshold} hours ({TAD_threshold_days} days)")

# ============================================================================
# Define simulation parameters and initial conditions
# Energy consumption parameters (kWh per unit production)
e_elz_h2          = 49    # electrolyzer energy demand for H2 production
e_elz_nh3         = 8.71  # electrolyzer energy demand for NH3 production
e_hb              = 0.6   # Haber-Bosch unit energy demand for NH3 production
e_psa             = 0.119 # energy demand for air separation (N2 production)
e_syngas_compression = 0.16 # energy for pre-compression of syngas
e_as = e_hb + e_psa + e_syngas_compression  # total for continuous ammonia synthesis
h2O_demand = 10  # liters H2O per kg H2 produced

# Production targets and derived quantities
nh3_prod_day         = 1500                        # tonnes NH3 produced per day
tonnes_nh3_prod_year = nh3_prod_day * (365 - TAD_threshold_days)
tonnes_nh3_prod_year_real = nh3_prod_day * 335      # alternative realistic production value
kg_nh3_prod_year     = tonnes_nh3_prod_year * 1000   # convert tonnes to kg
kg_nh3_prod_hour     = kg_nh3_prod_year / 8760       # hourly production (kgNH3/h)

# Hydrogen and nitrogen demands for ammonia synthesis (example factors)
m_h2_asu_year = kg_nh3_prod_year * 0.1777  # yearly H2 demand (kgH2)
m_h2_asu_hour = kg_nh3_prod_hour * 0.1777  # hourly H2 demand (kgH2)
m_n2_asu_year = kg_nh3_prod_year * 0.8223  # yearly N2 demand (kgN2)
m_n2_asu_hour = kg_nh3_prod_hour * 0.8223  # hourly N2 demand (kgN2)

# Installed capacities (kW)
p_hb_installed     = kg_nh3_prod_hour * e_hb
p_psa_installed    = m_n2_asu_hour * e_psa
p_syngas_compression = kg_nh3_prod_hour * e_syngas_compression
p_asu_installed    = p_hb_installed + p_psa_installed + p_syngas_compression
p_elz_installed    = m_h2_asu_hour * e_elz_h2
p_total            = p_elz_installed + p_asu_installed  # total required capacity

# ============================================================================
# Define installed capacity ranges for system components
# (First, based on annual demand)
wt_step   = max(p_total * 0.05, 0.05)
pv_step   = max(p_total * 0.05, 0.05)
elz_step  = max(p_elz_installed * 0.03, 0.5)
bat_step  = max(p_asu_installed * 6, 0.5)
h2tank_step = max(m_h2_asu_hour * 12, 0.1)

# ADJUSTED: Increased capacities to better match renewable resources
wt_min, wt_max   = (p_total * 2.0), (p_total * 4.0)
pv_min, pv_max   = (p_total * 0.5), (p_total * 1.0)
elz_min, elz_max = (p_elz_installed * 1.5), (p_elz_installed * 2.0)
bat_min, bat_max = (p_asu_installed * 2), (p_asu_installed * 48)
h2tank_min, h2tank_max = (m_h2_asu_hour * (24 * 5)), (m_h2_asu_hour * (24 * 14))

wt_range    = np.arange(wt_min, wt_max + wt_step, wt_step)
pv_range    = np.arange(pv_min, pv_max + pv_step, pv_step)
elz_range   = np.arange(elz_min, elz_max + elz_step, elz_step)
bat_range   = np.arange(bat_min, bat_max + bat_step, bat_step)
h2tank_range= np.arange(h2tank_min, h2tank_max + h2tank_step, h2tank_step)

# Rather than using refined ranges, keep the ranges broader for testing
print(f"Wind range: {wt_min:.0f} - {wt_max:.0f} kW, step {wt_step:.0f} kW")
print(f"PV range: {pv_min:.0f} - {pv_max:.0f} kW, step {pv_step:.0f} kW")
print(f"Electrolyzer range: {elz_min:.0f} - {elz_max:.0f} kW, step {elz_step:.0f} kW")
print(f"Battery range: {bat_min:.0f} - {bat_max:.0f} kWh, step {bat_step:.0f} kWh")
print(f"H2 tank range: {h2tank_min:.0f} - {h2tank_max:.0f} kgH2, step {h2tank_step:.0f} kgH2")

# ============================================================================
# Preallocate arrays for simulation outputs
# ============================================================================
p_wt       = np.zeros(t_sim)  # wind turbine power output
p_pv       = np.zeros(t_sim)  # PV power output
p_re       = np.zeros(t_sim)  # total renewable power
p_tot      = np.zeros(t_sim)  # power available after ASU consumption
c_h2tank   = np.zeros(t_sim)  # hydrogen tank state (kgH2)
c_bat      = np.zeros(t_sim)  # battery state (kWh)
p_elz      = np.zeros(t_sim)  # electrolyzer power consumption
p_bat_in   = np.zeros(t_sim)  # battery charging power
p_bat_out  = np.zeros(t_sim)  # battery discharging power
p_asu      = np.zeros(t_sim)  # power consumed by ammonia synthesis unit
m_h2_elz   = np.zeros(t_sim)  # hydrogen produced by electrolyzers (kgH2/h)
m_h2_asu   = np.zeros(t_sim)  # hydrogen consumed by ASU (kgH2/h)
m_h2tank_in  = np.zeros(t_sim)  # hydrogen charged into tank (kgH2/h)
m_h2tank_out = np.zeros(t_sim)  # hydrogen discharged from tank (kgH2/h)
energy_curt  = np.zeros(t_sim)  # curtailed energy (kWh)

simulation_results = []

# ============================================================================
# Define functions to calculate CAPEX, OPEX, LCOA and electrolyzer constraints
# ============================================================================
def calculate_capex(wt, pv, elz, bat, h2tank, p_hb_installed, p_syngas_compression, p_psa_installed):
    capex_wt   = 1150 * wt
    capex_pv   = 475 * pv
    capex_bat  = 250 * bat
    capex_elz  = 610 * elz
    capex_h2tank = 400 * h2tank
    capex_asu  = 2837 * (p_hb_installed + p_syngas_compression)
    capex_psa  = 3490 * p_psa_installed
    return np.array([capex_wt, capex_pv, capex_bat, capex_elz, capex_h2tank, capex_asu, capex_psa])

def calculate_opex(capex):
    opex_wt   = 0.02 * capex[0]
    opex_pv   = 0.02 * capex[1]
    opex_bat  = 0.01 * capex[2]
    opex_elz  = 0.03 * capex[3]
    opex_h2tank = 0 * capex[4]
    opex_asu  = 0.05 * capex[5]
    opex_psa  = 0.02 * capex[6]
    return np.array([opex_wt, opex_pv, opex_bat, opex_elz, opex_h2tank, opex_asu, opex_psa])

def calculate_lcoa(capex, opex, annual_ammonia_prod):
    plant_lifetime = 30  # years
    discount_rate  = 0.075
    # Annualized capital cost for each component
    ann_capex = (capex * (discount_rate * (1 + discount_rate)**plant_lifetime)
                 / ((1 + discount_rate)**plant_lifetime - 1))
    tac = ann_capex + opex  # total annual cost
    return np.sum(tac) / annual_ammonia_prod

def calculate_elz_constraints(elz):
    p_elz_max = elz * 1.15
    p_elz_min = elz * 0.15
    return p_elz_max, p_elz_min

# ============================================================================
# Complete the simulation logic with power and hydrogen flows
# ============================================================================
# FIXED: Changed function signature to receive power_value instead of p_tot array
def simulate_power_flows(power_value, p_elz_max, p_elz_min, c_bat_array, c_h2tank_array,
                         c_bat_max, c_bat_min, c_h2tank_max, c_h2tank_min, m_h2_asu_hour, t):
    # Initialize default values
    p_elz[t] = 0
    p_bat_in[t] = 0
    p_bat_out[t] = 0
    p_asu[t] = p_asu_installed
    m_h2_elz[t] = 0
    m_h2_asu[t] = 0
    m_h2tank_in[t] = 0
    m_h2tank_out[t] = 0
    energy_curt[t] = 0

    # Get current states
    c_bat_current = c_bat_array[t]
    c_h2tank_current = c_h2tank_array[t]

    # Case 1: Power deficit scenario
    if power_value < 0:
        # Try to cover deficit from battery
        if c_bat_current > c_bat_min:
            p_bat_out[t] = min(abs(power_value), 0.5 * (c_bat_current - c_bat_min))
            c_bat_array[t] = c_bat_current - p_bat_out[t]

        # If battery is insufficient, draw from H2 tank
        remaining_deficit = abs(power_value) - p_bat_out[t]
        if remaining_deficit > 0 and c_h2tank_current > c_h2tank_min:
            m_h2tank_out[t] = min(m_h2_asu_hour, (c_h2tank_current - c_h2tank_min))
            m_h2_asu[t] = m_h2tank_out[t]
            c_h2tank_array[t] = c_h2tank_current - m_h2tank_out[t]
        else:
            m_h2_asu[t] = 0  # No hydrogen available

    # Case 2: Just enough power for ASU but not for electrolyzers
    elif 0 <= power_value < p_elz_min:
        # Use hydrogen from tank
        if c_h2tank_current > c_h2tank_min:
            m_h2tank_out[t] = min(m_h2_asu_hour, (c_h2tank_current - c_h2tank_min))
            m_h2_asu[t] = m_h2tank_out[t]
            c_h2tank_array[t] = c_h2tank_current - m_h2tank_out[t]

            # Charge battery with remaining power if possible
            if power_value > 0 and c_bat_current < c_bat_max:
                p_bat_in[t] = min(power_value, 0.5 * (c_bat_max - c_bat_current))
                c_bat_array[t] = c_bat_current + p_bat_in[t] * 0.9  # 90% charging efficiency
        else:
            m_h2_asu[t] = 0  # No hydrogen available

    # Case 3: Electrolyzer can operate
    elif p_elz_min <= power_value <= p_elz_max:
        # Run electrolyzer with available power
        p_elz[t] = power_value
        m_h2_elz[t] = p_elz[t] / e_elz_h2

        # Determine how much H2 goes to ASU vs tank
        if m_h2_elz[t] < m_h2_asu_hour:  # Not enough for ASU
            # Draw additional H2 from tank if needed
            if c_h2tank_current > c_h2tank_min:
                m_h2tank_out[t] = min(m_h2_asu_hour - m_h2_elz[t],
                                     (c_h2tank_current - c_h2tank_min))
                m_h2_asu[t] = m_h2_elz[t] + m_h2tank_out[t]
                c_h2tank_array[t] = c_h2tank_current - m_h2tank_out[t]
            else:
                m_h2_asu[t] = m_h2_elz[t]  # Use what we have
        else:  # Excess H2 production
            m_h2_asu[t] = m_h2_asu_hour  # ASU gets what it needs
            # Store excess H2 in tank
            if c_h2tank_current < c_h2tank_max:
                m_h2tank_in[t] = min(m_h2_elz[t] - m_h2_asu_hour,
                                    (c_h2tank_max - c_h2tank_current))
                c_h2tank_array[t] = c_h2tank_current + m_h2tank_in[t]
            else:
                # If tank is full, this H2 is effectively curtailed
                energy_curt[t] += (m_h2_elz[t] - m_h2_asu_hour - m_h2tank_in[t]) * e_elz_h2

    # Case 4: Surplus power beyond electrolyzer capacity
    elif power_value > p_elz_max:
        # Run electrolyzer at max capacity
        p_elz[t] = p_elz_max
        m_h2_elz[t] = p_elz[t] / e_elz_h2
        m_h2_asu[t] = m_h2_asu_hour  # ASU gets what it needs

        # Store excess H2 in tank
        if c_h2tank_current < c_h2tank_max:
            m_h2tank_in[t] = min(m_h2_elz[t] - m_h2_asu_hour,
                                (c_h2tank_max - c_h2tank_current))
            c_h2tank_array[t] = c_h2tank_current + m_h2tank_in[t]

        # Use remaining surplus to charge battery
        surplus = power_value - p_elz_max
        if surplus > 0 and c_bat_current < c_bat_max:
            p_bat_in[t] = min(surplus, 0.5 * (c_bat_max - c_bat_current))
            c_bat_array[t] = c_bat_current + p_bat_in[t] * 0.9  # 90% charging efficiency
            surplus -= p_bat_in[t]

        # Any remaining power is curtailed
        if surplus > 0:
            energy_curt[t] += surplus

    return m_h2_asu[t]

# ============================================================================
# Simulation: Loop over all configurations and record the results
# ============================================================================
min_lcoa = float('inf')
best_config = None
best_results = None

# For quick testing, reduce the range sizes (can be removed for full simulation)
wt_range = wt_range[::3]  # Take every 3rd value
pv_range = pv_range[::3]  # Take every 3rd value
elz_range = elz_range[::2]  # Take every 2nd value
bat_range = bat_range[::3]  # Take every 3rd value
h2tank_range = h2tank_range[::3]  # Take every 3rd value

# ADDED: Fallback in case no configuration meets TAD threshold
min_tad = float('inf')
min_tad_config = None
min_tad_results = None

print(f"Starting simulation with {len(wt_range)}{len(pv_range)}{len(elz_range)}{len(bat_range)}{len(h2tank_range)} configurations")

for wt in wt_range:  # Installed wind capacity (kW)
    for pv in pv_range:  # Installed PV capacity (kW)
        for elz in elz_range:  # Installed electrolyzer capacity (kW)
            p_elz_max, p_elz_min = calculate_elz_constraints(elz)
            for bat in bat_range:  # Battery capacity (kWh)
                for h2tank in h2tank_range:  # Hydrogen tank capacity (kgH2)
                    # Reset simulation data for each configuration
                    TAD = np.zeros(t_sim)
                    total_TAD = 0

                    # Reset state arrays for this configuration
                    p_wt.fill(0)
                    p_pv.fill(0)
                    p_re.fill(0)
                    p_tot.fill(0)
                    c_h2tank.fill(0)
                    c_bat.fill(0)
                    p_elz.fill(0)
                    p_bat_in.fill(0)
                    p_bat_out.fill(0)
                    p_asu.fill(0)
                    m_h2_elz.fill(0)
                    m_h2_asu.fill(0)
                    m_h2tank_in.fill(0)
                    m_h2tank_out.fill(0)
                    energy_curt.fill(0)

                    # Set operational limits for battery and H2 tank
                    c_bat_max = bat
                    c_bat_min = 0.2 * bat  # 20% minimum state of charge
                    c_h2tank_max = h2tank
                    c_h2tank_min = 0

                    # Initialize initial states (start with full storage)
                    c_bat[0] = c_bat_max
                    c_h2tank[0] = c_h2tank_max

                    print(f"Simulating: wt={wt:.1f}, pv={pv:.1f}, elz={elz:.1f}, bat={bat:.1f}, h2tank={h2tank:.1f}")

                    for t in range(t_sim):
                        # Calculate renewable generation scaled to installed capacities
                        p_wt[t] = wt * wind_power_1kwp[t]
                        p_pv[t] = pv * solar_power_1kwp[t]
                        p_re[t] = p_wt[t] + p_pv[t]
                        p_tot[t] = p_re[t] - p_asu_installed  # after ASU demand

                        # Perform power flow simulation with the current value of p_tot (not the array)
                        supplied_h2 = simulate_power_flows(
                            p_tot[t], p_elz_max, p_elz_min,
                            c_bat, c_h2tank, c_bat_max, c_bat_min,
                            c_h2tank_max, c_h2tank_min, m_h2_asu_hour, t
                        )

                        # Compute TAD (Total Ammonia Deficit) for the hour
                        TAD[t] = 1 if supplied_h2 < m_h2_asu_hour else 0

                    # Calculate total deficit hours and percentage
                    total_TAD = np.sum(TAD)
                    tad_percentage = (total_TAD / t_sim) * 100

                    print(f"Total TAD = {total_TAD} hours ({tad_percentage:.2f}%)")

                    # Keep track of min TAD configuration as a fallback
                    if total_TAD < min_tad:
                        min_tad = total_TAD

                        # Calculate key metrics
                        capex = calculate_capex(wt, pv, elz, bat, h2tank,
                                            p_hb_installed, p_syngas_compression, p_psa_installed)
                        opex = calculate_opex(capex)
                        lcoa = calculate_lcoa(capex, opex, tonnes_nh3_prod_year_real)

                        total_capex = np.sum(capex)
                        total_opex = np.sum(opex)

                        min_tad_config = {
                            'wt': wt,
                            'pv': pv,
                            'elz': elz,
                            'bat': bat,
                            'h2tank': h2tank,
                            'lcoa': lcoa,
                            'capex': total_capex,
                            'opex': total_opex,
                            'tad': total_TAD
                        }

                        min_tad_results = {
                            'renewable_generation': np.sum(p_re),
                            'curtailed_energy': np.sum(energy_curt),
                            'h2_produced': np.sum(m_h2_elz),
                            'h2_consumed': np.sum(m_h2_asu),
                            'battery_cycles': np.sum(p_bat_in) / bat if bat > 0 else 0,
                            'h2_tank_cycles': np.sum(m_h2tank_in) / h2tank if h2tank > 0 else 0
                        }

                    if total_TAD <= TAD_threshold:
                        # Calculate CAPEX, OPEX, and LCOA for this configuration
                        capex = calculate_capex(wt, pv, elz, bat, h2tank,
                                              p_hb_installed, p_syngas_compression, p_psa_installed)
                        opex = calculate_opex(capex)
                        lcoa = calculate_lcoa(capex, opex, tonnes_nh3_prod_year_real)

                        total_capex = np.sum(capex)
                        total_opex = np.sum(opex)

                        print(f"LCOA = {lcoa:.2f} USD/ton NH3, CAPEX = {total_capex/1e6:.2f} MUSD")

                        if lcoa < min_lcoa:
                            min_lcoa = lcoa
                            best_config = {
                                'wt': wt,
                                'pv': pv,
                                'elz': elz,
                                'bat': bat,
                                'h2tank': h2tank,
                                'lcoa': lcoa,
                                'capex': total_capex,
                                'opex': total_opex,
                                'tad': total_TAD
                            }

                            # Save key performance indicators for the best configuration
                            best_results = {
                                'renewable_generation': np.sum(p_re),
                                'curtailed_energy': np.sum(energy_curt),
                                'h2_produced': np.sum(m_h2_elz),
                                'h2_consumed': np.sum(m_h2_asu),
                                'battery_cycles': np.sum(p_bat_in) / bat if bat > 0 else 0,
                                'h2_tank_cycles': np.sum(m_h2tank_in) / h2tank if h2tank > 0 else 0
                            }
                    else:
                        continue  # TAD exceeds threshold; skip this configuration

# ADDED: If no configuration meets the TAD threshold, use the one with minimum TAD
if best_config is None:
    print("\nNo configuration met the TAD threshold. Using configuration with minimum TAD.")
    best_config = min_tad_config
    best_results = min_tad_results

print("\n=== OPTIMIZATION RESULTS ===")
print("Best Configuration:")
for key, value in best_config.items():
    print(f"  {key}: {value}")

print("\nPerformance Metrics:")
for key, value in best_results.items():
    print(f"  {key}: {value}")

# Save the best configuration results to a file
with open('optimization_results.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Parameter', 'Value'])
    for key, value in best_config.items():
        writer.writerow([key, value])
    for key, value in best_results.items():
        writer.writerow([key, value])
```

```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt

# ============================================================================
# Choose the data source for the renewable production data (now using ninja)
# ============================================================================
data_source = "ninja"  # Using ninja datasets

# Update the file paths to the ninja datasets
solar_file = "ninja_pv_country_DK_merra-2_nuts-2_corrected.csv"
wind_file = "ninja_wind_country_DK_current_merra-2_nuts-2_corrected.csv"

# Read the CSV files, skipping the first two rows which contain metadata
# For PV data
solar_data = []
with open(solar_file, 'r') as f:
    lines = f.readlines()
    pv_headers = lines[2].strip().split(',')
    for line in lines[3:]:
        if line.strip() and line.startswith('2018'):  # Using 2018 data
            values = line.strip().split(',')
            solar_data.append(values)

# For wind data
wind_data = []
with open(wind_file, 'r') as f:
    lines = f.readlines()
    wind_headers = lines[2].strip().split(',')
    for line in lines[3:]:
        if line.strip() and line.startswith('2018'):  # Using 2018 data
            values = line.strip().split(',')
            wind_data.append(values)

# Extract the capacity factors - using DK01 for PV and DK_TOTAL for wind if available
pv_column_idx = pv_headers.index('DK01')
wind_column_idx = wind_headers.index('DK_TOTAL') if 'DK_TOTAL' in wind_headers else wind_headers.index('DK01')

# Extract hourly production data
solar_power_1kwp = [float(row[pv_column_idx]) for row in solar_data][:8760]
wind_power_1kwp = [float(row[wind_column_idx]) for row in wind_data][:8760]

# ============================================================================
# Simulation time settings
# ============================================================================
dt    = 1      # time step in hours
t_sim = 8760   # simulation period (1 year)

# TAD threshold (Total Ammonia Deficit) to limit iterations
TAD_threshold      = 24 * 300       # 300 days worth of hours (changed from 5 days)
TAD_threshold_days = TAD_threshold / 24

print(f"Using adjusted TAD threshold of {TAD_threshold} hours ({TAD_threshold_days} days)")

# ============================================================================
# Define simulation parameters and initial conditions
# ============================================================================
# Energy consumption parameters (kWh per unit production)
e_elz_h2          = 49    # electrolyzer energy demand for H2 production
e_elz_nh3         = 8.71  # electrolyzer energy demand for NH3 production
e_hb              = 0.6   # Haber-Bosch unit energy demand for NH3 production
e_psa             = 0.119 # energy demand for air separation (N2 production)
e_syngas_compression = 0.16 # energy for pre-compression of syngas
e_as = e_hb + e_psa + e_syngas_compression  # total for continuous ammonia synthesis
h2O_demand = 10  # liters H2O per kg H2 produced

# Production targets and derived quantities
nh3_prod_day         = 1500                        # tonnes NH3 produced per day
tonnes_nh3_prod_year = nh3_prod_day * (365 - TAD_threshold_days)
tonnes_nh3_prod_year_real = nh3_prod_day * 335      # alternative realistic production value
kg_nh3_prod_year     = tonnes_nh3_prod_year * 1000   # convert tonnes to kg
kg_nh3_prod_hour     = kg_nh3_prod_year / 8760       # hourly production (kgNH3/h)

# Hydrogen and nitrogen demands for ammonia synthesis (example factors)
m_h2_asu_year = kg_nh3_prod_year * 0.1777  # yearly H2 demand (kgH2)
m_h2_asu_hour = kg_nh3_prod_hour * 0.1777  # hourly H2 demand (kgH2)
m_n2_asu_year = kg_nh3_prod_year * 0.8223  # yearly N2 demand (kgN2)
m_n2_asu_hour = kg_nh3_prod_hour * 0.8223  # hourly N2 demand (kgN2)

# Installed capacities (kW)
p_hb_installed     = kg_nh3_prod_hour * e_hb
p_psa_installed    = m_n2_asu_hour * e_psa
p_syngas_compression = kg_nh3_prod_hour * e_syngas_compression
p_asu_installed    = p_hb_installed + p_psa_installed + p_syngas_compression
p_elz_installed    = m_h2_asu_hour * e_elz_h2
p_total            = p_elz_installed + p_asu_installed  # total required capacity

# ============================================================================
# Define installed capacity ranges for system components
# ============================================================================
wt_step   = max(p_total * 0.05, 0.05)
pv_step   = max(p_total * 0.05, 0.05)
elz_step  = max(p_elz_installed * 0.03, 0.5)
bat_step  = max(p_asu_installed * 6, 0.5)
h2tank_step = max(m_h2_asu_hour * 12, 0.1)

# ADJUSTED: Increased capacities to better match renewable resources
wt_min, wt_max   = (p_total * 2.0), (p_total * 4.0)
pv_min, pv_max   = (p_total * 0.5), (p_total * 1.0)
elz_min, elz_max = (p_elz_installed * 1.5), (p_elz_installed * 2.0)
bat_min, bat_max = (p_asu_installed * 2), (p_asu_installed * 48)
h2tank_min, h2tank_max = (m_h2_asu_hour * (24 * 5)), (m_h2_asu_hour * (24 * 14))

wt_range    = np.arange(wt_min, wt_max + wt_step, wt_step)
pv_range    = np.arange(pv_min, pv_max + pv_step, pv_step)
elz_range   = np.arange(elz_min, elz_max + elz_step, elz_step)
bat_range   = np.arange(bat_min, bat_max + bat_step, bat_step)
h2tank_range= np.arange(h2tank_min, h2tank_max + h2tank_step, h2tank_step)

# Rather than using refined ranges, keep the ranges broader for testing
print(f"Wind range: {wt_min:.0f} - {wt_max:.0f} kW, step {wt_step:.0f} kW")
print(f"PV range: {pv_min:.0f} - {pv_max:.0f} kW, step {pv_step:.0f} kW")
print(f"Electrolyzer range: {elz_min:.0f} - {elz_max:.0f} kW, step {elz_step:.0f} kW")
print(f"Battery range: {bat_min:.0f} - {bat_max:.0f} kWh, step {bat_step:.0f} kWh")
print(f"H2 tank range: {h2tank_min:.0f} - {h2tank_max:.0f} kgH2, step {h2tank_step:.0f} kgH2")

# ============================================================================
# Preallocate arrays for simulation outputs
# ============================================================================
p_wt       = np.zeros(t_sim)  # wind turbine power output
p_pv       = np.zeros(t_sim)  # PV power output
p_re       = np.zeros(t_sim)  # total renewable power
p_tot      = np.zeros(t_sim)  # power available after ASU consumption
c_h2tank   = np.zeros(t_sim)  # hydrogen tank state (kgH2)
c_bat      = np.zeros(t_sim)  # battery state (kWh)
p_elz      = np.zeros(t_sim)  # electrolyzer power consumption
p_bat_in   = np.zeros(t_sim)  # battery charging power
p_bat_out  = np.zeros(t_sim)  # battery discharging power
p_asu      = np.zeros(t_sim)  # power consumed by ammonia synthesis unit
m_h2_elz   = np.zeros(t_sim)  # hydrogen produced by electrolyzers (kgH2/h)
m_h2_asu   = np.zeros(t_sim)  # hydrogen consumed by ASU (kgH2/h)
m_h2tank_in  = np.zeros(t_sim)  # hydrogen charged into tank (kgH2/h)
m_h2tank_out = np.zeros(t_sim)  # hydrogen discharged from tank (kgH2/h)
energy_curt  = np.zeros(t_sim)  # curtailed energy (kWh)

simulation_results = []

# ============================================================================
# Define functions to calculate CAPEX, OPEX, LCOA and electrolyzer constraints
# ============================================================================
def calculate_capex(wt, pv, elz, bat, h2tank, p_hb_installed, p_syngas_compression, p_psa_installed):
    capex_wt   = 1150 * wt
    capex_pv   = 475 * pv
    capex_bat  = 250 * bat
    capex_elz  = 610 * elz
    capex_h2tank = 400 * h2tank
    capex_asu  = 2837 * (p_hb_installed + p_syngas_compression)
    capex_psa  = 3490 * p_psa_installed
    return np.array([capex_wt, capex_pv, capex_bat, capex_elz, capex_h2tank, capex_asu, capex_psa])

def calculate_opex(capex):
    opex_wt   = 0.02 * capex[0]
    opex_pv   = 0.02 * capex[1]
    opex_bat  = 0.01 * capex[2]
    opex_elz  = 0.03 * capex[3]
    opex_h2tank = 0 * capex[4]
    opex_asu  = 0.05 * capex[5]
    opex_psa  = 0.02 * capex[6]
    return np.array([opex_wt, opex_pv, opex_bat, opex_elz, opex_h2tank, opex_asu, opex_psa])

def calculate_lcoa(capex, opex, annual_ammonia_prod):
    plant_lifetime = 30  # years
    discount_rate  = 0.075
    # Annualized capital cost for each component
    ann_capex = (capex * (discount_rate * (1 + discount_rate)**plant_lifetime)
                 / ((1 + discount_rate)**plant_lifetime - 1))
    tac = ann_capex + opex  # total annual cost
    return np.sum(tac) / annual_ammonia_prod

def calculate_elz_constraints(elz):
    p_elz_max = elz * 1.15
    p_elz_min = elz * 0.15
    return p_elz_max, p_elz_min

# ============================================================================
# Complete the simulation logic with power and hydrogen flows
# ============================================================================
def simulate_power_flows(power_value, p_elz_max, p_elz_min, c_bat_array, c_h2tank_array,
                         c_bat_max, c_bat_min, c_h2tank_max, c_h2tank_min, m_h2_asu_hour, t):
    # Initialize default values
    p_elz[t] = 0
    p_bat_in[t] = 0
    p_bat_out[t] = 0
    p_asu[t] = p_asu_installed
    m_h2_elz[t] = 0
    m_h2_asu[t] = 0
    m_h2tank_in[t] = 0
    m_h2tank_out[t] = 0
    energy_curt[t] = 0

    # Get current states
    c_bat_current = c_bat_array[t]
    c_h2tank_current = c_h2tank_array[t]

    # Case 1: Power deficit scenario
    if power_value < 0:
        # Try to cover deficit from battery
        if c_bat_current > c_bat_min:
            p_bat_out[t] = min(abs(power_value), 0.5 * (c_bat_current - c_bat_min))
            c_bat_array[t] = c_bat_current - p_bat_out[t]

        # If battery is insufficient, draw from H2 tank
        remaining_deficit = abs(power_value) - p_bat_out[t]
        if remaining_deficit > 0 and c_h2tank_current > c_h2tank_min:
            m_h2tank_out[t] = min(m_h2_asu_hour, (c_h2tank_current - c_h2tank_min))
            m_h2_asu[t] = m_h2tank_out[t]
            c_h2tank_array[t] = c_h2tank_current - m_h2tank_out[t]
        else:
            m_h2_asu[t] = 0  # No hydrogen available

    # Case 2: Just enough power for ASU but not for electrolyzers
    elif 0 <= power_value < p_elz_min:
        # Use hydrogen from tank
        if c_h2tank_current > c_h2tank_min:
            m_h2tank_out[t] = min(m_h2_asu_hour, (c_h2tank_current - c_h2tank_min))
            m_h2_asu[t] = m_h2tank_out[t]
            c_h2tank_array[t] = c_h2tank_current - m_h2tank_out[t]

            # Charge battery with remaining power if possible
            if power_value > 0 and c_bat_current < c_bat_max:
                p_bat_in[t] = min(power_value, 0.5 * (c_bat_max - c_bat_current))
                c_bat_array[t] = c_bat_current + p_bat_in[t] * 0.9  # 90% charging efficiency
        else:
            m_h2_asu[t] = 0  # No hydrogen available

    # Case 3: Electrolyzer can operate
    elif p_elz_min <= power_value <= p_elz_max:
        # Run electrolyzer with available power
        p_elz[t] = power_value
        m_h2_elz[t] = p_elz[t] / e_elz_h2

        # Determine how much H2 goes to ASU vs tank
        if m_h2_elz[t] < m_h2_asu_hour:  # Not enough for ASU
            # Draw additional H2 from tank if needed
            if c_h2tank_current > c_h2tank_min:
                m_h2tank_out[t] = min(m_h2_asu_hour - m_h2_elz[t],
                                     (c_h2tank_current - c_h2tank_min))
                m_h2_asu[t] = m_h2_elz[t] + m_h2tank_out[t]
                c_h2tank_array[t] = c_h2tank_current - m_h2tank_out[t]
            else:
                m_h2_asu[t] = m_h2_elz[t]  # Use what we have
        else:  # Excess H2 production
            m_h2_asu[t] = m_h2_asu_hour  # ASU gets what it needs
            # Store excess H2 in tank
            if c_h2tank_current < c_h2tank_max:
                m_h2tank_in[t] = min(m_h2_elz[t] - m_h2_asu_hour,
                                    (c_h2tank_max - c_h2tank_current))
                c_h2tank_array[t] = c_h2tank_current + m_h2tank_in[t]
            else:
                # If tank is full, this H2 is effectively curtailed
                energy_curt[t] += (m_h2_elz[t] - m_h2_asu_hour - m_h2tank_in[t])

    # Case 4: Surplus power beyond electrolyzer capacity
    elif power_value > p_elz_max:
        # Run electrolyzer at max capacity
        p_elz[t] = p_elz_max
        m_h2_elz[t] = p_elz[t] / e_elz_h2
        m_h2_asu[t] = m_h2_asu_hour  # ASU gets what it needs

        # Store excess H2 in tank
        if c_h2tank_current < c_h2tank_max:
            m_h2tank_in[t] = min(m_h2_elz[t] - m_h2_asu_hour,
                                (c_h2tank_max - c_h2tank_current))
            c_h2tank_array[t] = c_h2tank_current + m_h2tank_in[t]

        # Use remaining surplus to charge battery
        surplus = power_value - p_elz_max
        if surplus > 0 and c_bat_current < c_bat_max:
            p_bat_in[t] = min(surplus, 0.5 * (c_bat_max - c_bat_current))
            c_bat_array[t] = c_bat_current + p_bat_in[t] * 0.9  # 90% charging efficiency
            surplus -= p_bat_in[t]

        # Any remaining power is curtailed
        if surplus > 0:
            energy_curt[t] += surplus

    return m_h2_asu[t]

# ============================================================================
# Simulation: Loop over all configurations and record the results
# ============================================================================
min_lcoa = float('inf')
best_config = None
best_results = None

# For quick testing, reduce the range sizes (can be removed for full simulation)
wt_range = wt_range[::3]  # Take every 3rd value
pv_range = pv_range[::3]  # Take every 3rd value
elz_range = elz_range[::2]  # Take every 2nd value
bat_range = bat_range[::3]  # Take every 3rd value
h2tank_range = h2tank_range[::3]  # Take every 3rd value

# ADDED: Fallback in case no configuration meets TAD threshold
min_tad = float('inf')
min_tad_config = None
min_tad_results = None

# Variable to store simulation arrays for the best configuration (for plotting)
best_simulation_data = None

print(f"Starting simulation with {len(wt_range)}{len(pv_range)}{len(elz_range)}{len(bat_range)}{len(h2tank_range)} configurations")

for wt in wt_range:  # Installed wind capacity (kW)
    for pv in pv_range:  # Installed PV capacity (kW)
        for elz in elz_range:  # Installed electrolyzer capacity (kW)
            p_elz_max, p_elz_min = calculate_elz_constraints(elz)
            for bat in bat_range:  # Battery capacity (kWh)
                for h2tank in h2tank_range:  # Hydrogen tank capacity (kgH2)
                    # Reset simulation data for each configuration
                    TAD = np.zeros(t_sim)
                    total_TAD = 0

                    # Reset state arrays for this configuration
                    p_wt.fill(0)
                    p_pv.fill(0)
                    p_re.fill(0)
                    p_tot.fill(0)
                    c_h2tank.fill(0)
                    c_bat.fill(0)
                    p_elz.fill(0)
                    p_bat_in.fill(0)
                    p_bat_out.fill(0)
                    p_asu.fill(0)
                    m_h2_elz.fill(0)
                    m_h2_asu.fill(0)
                    m_h2tank_in.fill(0)
                    m_h2tank_out.fill(0)
                    energy_curt.fill(0)

                    # Set operational limits for battery and H2 tank
                    c_bat_max = bat
                    c_bat_min = 0.2 * bat  # 20% minimum state of charge
                    c_h2tank_max = h2tank
                    c_h2tank_min = 0

                    # Initialize initial states (start with full storage)
                    c_bat[0] = c_bat_max
                    c_h2tank[0] = c_h2tank_max

                    print(f"Simulating: wt={wt:.1f}, pv={pv:.1f}, elz={elz:.1f}, bat={bat:.1f}, h2tank={h2tank:.1f}")

                    for t in range(t_sim):
                        # Calculate renewable generation scaled to installed capacities
                        p_wt[t] = wt * wind_power_1kwp[t]
                        p_pv[t] = pv * solar_power_1kwp[t]
                        p_re[t] = p_wt[t] + p_pv[t]
                        p_tot[t] = p_re[t] - p_asu_installed  # after ASU demand

                        # Perform power flow simulation with the current value of p_tot (not the array)
                        supplied_h2 = simulate_power_flows(
                            p_tot[t], p_elz_max, p_elz_min,
                            c_bat, c_h2tank, c_bat_max, c_bat_min,
                            c_h2tank_max, c_h2tank_min, m_h2_asu_hour, t
                        )

                        # Compute TAD (Total Ammonia Deficit) for the hour
                        TAD[t] = 1 if supplied_h2 < m_h2_asu_hour else 0

                    # Calculate total deficit hours and percentage
                    total_TAD = np.sum(TAD)
                    tad_percentage = (total_TAD / t_sim) * 100

                    print(f"Total TAD = {total_TAD} hours ({tad_percentage:.2f}%)")

                    # Keep track of min TAD configuration as a fallback
                    if total_TAD < min_tad:
                        min_tad = total_TAD

                        # Calculate key metrics
                        capex = calculate_capex(wt, pv, elz, bat, h2tank,
                                            p_hb_installed, p_syngas_compression, p_psa_installed)
                        opex = calculate_opex(capex)
                        lcoa = calculate_lcoa(capex, opex, tonnes_nh3_prod_year_real)

                        total_capex = np.sum(capex)
                        total_opex = np.sum(opex)

                        min_tad_config = {
                            'wt': wt,
                            'pv': pv,
                            'elz': elz,
                            'bat': bat,
                            'h2tank': h2tank,
                            'lcoa': lcoa,
                            'capex': total_capex,
                            'opex': total_opex,
                            'tad': total_TAD
                        }

                        min_tad_results = {
                            'renewable_generation': np.sum(p_re),
                            'curtailed_energy': np.sum(energy_curt),
                            'h2_produced': np.sum(m_h2_elz),
                            'h2_consumed': np.sum(m_h2_asu),
                            'battery_cycles': np.sum(p_bat_in) / bat if bat > 0 else 0,
                            'h2_tank_cycles': np.sum(m_h2tank_in) / h2tank if h2tank > 0 else 0
                        }

                    if total_TAD <= TAD_threshold:
                        # Calculate CAPEX, OPEX, and LCOA for this configuration
                        capex = calculate_capex(wt, pv, elz, bat, h2tank,
                                              p_hb_installed, p_syngas_compression, p_psa_installed)
                        opex = calculate_opex(capex)
                        lcoa = calculate_lcoa(capex, opex, tonnes_nh3_prod_year_real)

                        total_capex = np.sum(capex)
                        total_opex = np.sum(opex)

                        print(f"LCOA = {lcoa:.2f} USD/ton NH3, CAPEX = {total_capex/1e6:.2f} MUSD")

                        if lcoa < min_lcoa:
                            min_lcoa = lcoa
                            best_config = {
                                'wt': wt,
                                'pv': pv,
                                'elz': elz,
                                'bat': bat,
                                'h2tank': h2tank,
                                'lcoa': lcoa,
                                'capex': total_capex,
                                'opex': total_opex,
                                'tad': total_TAD
                            }

                            # Save key performance indicators for the best configuration
                            best_results = {
                                'renewable_generation': np.sum(p_re),
                                'curtailed_energy': np.sum(energy_curt),
                                'h2_produced': np.sum(m_h2_elz),
                                'h2_consumed': np.sum(m_h2_asu),
                                'battery_cycles': np.sum(p_bat_in) / bat if bat > 0 else 0,
                                'h2_tank_cycles': np.sum(m_h2tank_in) / h2tank if h2tank > 0 else 0
                            }

                            # Store simulation arrays for plotting the best configuration
                            best_simulation_data = {
                                'p_wt': np.copy(p_wt),
                                'p_pv': np.copy(p_pv),
                                'p_re': np.copy(p_re),
                                'p_tot': np.copy(p_tot),
                                'p_elz': np.copy(p_elz),
                                'c_bat': np.copy(c_bat),
                                'c_h2tank': np.copy(c_h2tank),
                                'TAD': np.copy(TAD),
                                'energy_curt': np.copy(energy_curt)
                            }
                    else:
                        continue  # TAD exceeds threshold; skip this configuration

# ADDED: If no configuration meets the TAD threshold, use the one with minimum TAD
if best_config is None:
    print("\nNo configuration met the TAD threshold. Using configuration with minimum TAD.")
    best_config = min_tad_config
    best_results = min_tad_results

print("\n=== OPTIMIZATION RESULTS ===")
print("Best Configuration:")
for key, value in best_config.items():
    print(f"  {key}: {value}")

print("\nPerformance Metrics:")
for key, value in best_results.items():
    print(f"  {key}: {value}")

# Save the best configuration results to a file
with open('optimization_results.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Parameter', 'Value'])
    for key, value in best_config.items():
        writer.writerow([key, value])
    for key, value in best_results.items():
        writer.writerow([key, value])

# ============================================================================
# Plotting results for the best configuration (using a one-week time window)
# ============================================================================
if best_simulation_data is not None:
    hours_to_plot = 168  # Plotting the first week (168 hours)
    time = np.arange(hours_to_plot)

    # Figure 1: Renewable Generation, Electrolyzer Power, and Curtailed Energy
    plt.figure(figsize=(12,6))
    plt.plot(time, best_simulation_data['p_re'][:hours_to_plot], label='Total Renewable Generation (kW)')
    plt.plot(time, best_simulation_data['p_elz'][:hours_to_plot], label='Electrolyzer Power (kW)')
    plt.plot(time, best_simulation_data['energy_curt'][:hours_to_plot], label='Curtailed Energy (kWh)')
    plt.xlabel('Hour')
    plt.ylabel('Power / Energy')
    plt.title('Renewable Generation, Electrolyzer Power, and Curtailment (First Week)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Figure 2: Battery and Hydrogen Tank States
    plt.figure(figsize=(12,6))
    plt.plot(time, best_simulation_data['c_bat'][:hours_to_plot], label='Battery State (kWh)')
    plt.plot(time, best_simulation_data['c_h2tank'][:hours_to_plot], label='Hydrogen Tank Level (kgH2)')
    plt.xlabel('Hour')
    plt.ylabel('Storage Level')
    plt.title('Battery and Hydrogen Tank Storage States (First Week)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Figure 3: Wind vs PV Contributions (Stacked Area Plot)
    plt.figure(figsize=(12,6))
    plt.stackplot(time, best_simulation_data['p_wt'][:hours_to_plot], best_simulation_data['p_pv'][:hours_to_plot],
                  labels=['Wind Power (kW)', 'PV Power (kW)'])
    plt.xlabel('Hour')
    plt.ylabel('Power (kW)')
    plt.title('Wind vs PV Contributions (First Week)')
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Figure 4: TAD (Total Ammonia Deficit) Indicator
    plt.figure(figsize=(12,3))
    plt.step(time, best_simulation_data['TAD'][:hours_to_plot], where='post')
    plt.xlabel('Hour')
    plt.ylabel('TAD Indicator')
    plt.title('Total Ammonia Deficit (TAD) Indicator (First Week)')
    plt.ylim(-0.1, 1.1)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
