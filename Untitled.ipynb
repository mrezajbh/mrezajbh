```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import gurobipy as gp
from gurobipy import GRB
import time
import csv

# =============================================================================
# 1. Renewable Data Input
# =============================================================================
data_source = "ninja"
solar_file = "ninja_pv_country_DK_merra-2_nuts-2_corrected.csv"
wind_file = "ninja_wind_country_DK_current_merra-2_nuts-2_corrected.csv"

# Reading CSV files and extracting capacity factors for 2018 data
solar_data = []
with open(solar_file, 'r') as f:
    lines = f.readlines()
    pv_headers = lines[2].strip().split(',')
    for line in lines[3:]:
        if line.strip() and line.startswith('2018'):
            values = line.strip().split(',')
            solar_data.append(values)

wind_data = []
with open(wind_file, 'r') as f:
    lines = f.readlines()
    wind_headers = lines[2].strip().split(',')
    for line in lines[3:]:
        if line.strip() and line.startswith('2018'):
            values = line.strip().split(',')
            wind_data.append(values)

pv_column_idx = pv_headers.index('DK01')
wind_column_idx = wind_headers.index('DK_TOTAL') if 'DK_TOTAL' in wind_headers else wind_headers.index('DK01')
solar_power_1kwp = [float(row[pv_column_idx]) for row in solar_data][:8760]
wind_power_1kwp = [float(row[wind_column_idx]) for row in wind_data][:8760]

# =============================================================================
# 2. Simulation Time Settings & Thresholds
# =============================================================================
dt = 1              # time step in hours
t_sim = 8760        # full simulation period (1 year)
TAD_threshold = 24 * 300     # e.g. 300 days worth of hours deficit allowed
TAD_threshold_days = TAD_threshold / 24
print(f"Using adjusted TAD threshold of {TAD_threshold} hours ({TAD_threshold_days} days)")

# =============================================================================
# 3. Select Representative Days (4 typical days, one from each season)
# =============================================================================
# Select a day from each season (days start from 0)
winter_day = 15     # January 15
spring_day = 105    # April 15
summer_day = 196    # July 15
fall_day = 288      # October 15

typical_days = [winter_day, spring_day, summer_day, fall_day]
days_in_season = [90, 92, 92, 91]  # approximate days in each season

# Extract data for typical days (24 hours per day)
typical_day_data = []
for day in typical_days:
    start_hour = day * 24
    end_hour = start_hour + 24
    day_data = {
        'solar': solar_power_1kwp[start_hour:end_hour],
        'wind': wind_power_1kwp[start_hour:end_hour],
        'hours': list(range(start_hour, end_hour))
    }
    typical_day_data.append(day_data)

# Flatten the typical day data for the optimization model
typical_solar = []
typical_wind = []
hour_mapping = []  # Maps optimization time step to original hour in year

for i, day_data in enumerate(typical_day_data):
    typical_solar.extend(day_data['solar'])
    typical_wind.extend(day_data['wind'])
    hour_mapping.extend(day_data['hours'])

# Number of time periods in the reduced model (4 days x 24 hours)
time_periods = len(typical_solar)
print(f"Reduced model: {time_periods} hours (4 typical days)")

# Scaling factor to convert results from typical days to full year
# This is for calculating annual energy production, deficits, etc.
annual_scaling_factor = sum(days_in_season)  # should be close to 365

# Scale the deficit threshold for the reduced model
scaled_deficit_threshold = int(TAD_threshold * (time_periods / t_sim))
print(f"Scaled deficit threshold for typical days: {scaled_deficit_threshold} hours")

# =============================================================================
# 4. Process Parameters and Production Targets (For e-Methanol)
# =============================================================================
# Energy requirements
e_elz_h2 = 49      # kWh per kg H2
e_plasma_co = 60   # kWh per kg CO
e_methanol = 0.4   # kWh per kg methanol synthesized
e_syngas_compression = 0.16  # kWh per kg CH3OH equivalent

# Production target
methanol_prod_day = 1500  # tonnes of methanol produced per day
tonnes_methanol_prod_year = methanol_prod_day * (365 - TAD_threshold_days)
tonnes_methanol_prod_year_real = methanol_prod_day * 335  # alternative realistic value
kg_methanol_prod_year = tonnes_methanol_prod_year * 1000
kg_methanol_prod_hour = kg_methanol_prod_year / 8760

# Stoichiometry for methanol: CO + 2 H2 -> CH3OH
m_h2_req_hour = kg_methanol_prod_hour * 0.125  # kg H2 needed per hour
m_co_req_hour = kg_methanol_prod_hour * 0.875  # kg CO needed per hour

# =============================================================================
# 5. Installed Capacities Calculation
# =============================================================================
# Reference capacities for the conversion units:
p_elz_installed = m_h2_req_hour * e_elz_h2          # kW needed for H2 production
p_plasma_installed = m_co_req_hour * e_plasma_co     # kW needed for CO production
p_methanol_installed = kg_methanol_prod_hour * e_methanol   # kW for methanol synthesis

# Total conversion capacity and total power needed
p_conversion_total = p_elz_installed + p_plasma_installed
p_total = p_conversion_total + p_methanol_installed

# =============================================================================
# 6. Capacity Ranges for Optimization
# =============================================================================
# Define bounds for the decision variables
wt_min, wt_max = (p_total * 2.0), (p_total * 4.0)
pv_min, pv_max = (p_total * 0.5), (p_total * 1.0)
elz_min, elz_max = (p_elz_installed * 1.5), (p_elz_installed * 2.0)
plasma_min, plasma_max = (p_plasma_installed * 1.5), (p_plasma_installed * 2.0)
bat_min, bat_max = (p_methanol_installed * 2), (p_methanol_installed * 48)
h2tank_min, h2tank_max = (m_h2_req_hour * 24 * 5), (m_h2_req_hour * 24 * 14)
cotank_min, cotank_max = (m_co_req_hour * 24 * 5), (m_co_req_hour * 24 * 14)

print(f"Wind range: {wt_min:.0f} - {wt_max:.0f} kW")
print(f"PV range: {pv_min:.0f} - {pv_max:.0f} kW")
print(f"Electrolyzer range: {elz_min:.0f} - {elz_max:.0f} kW")
print(f"Plasma unit range: {plasma_min:.0f} - {plasma_max:.0f} kW")
print(f"Battery range: {bat_min:.0f} - {bat_max:.0f} kWh")
print(f"H2 tank range: {h2tank_min:.0f} - {h2tank_max:.0f} kg")
print(f"CO tank range: {cotank_min:.0f} - {cotank_max:.0f} kg")

# =============================================================================
# 7. Cost Model Parameters
# =============================================================================
# Unit costs for CAPEX calculations
unit_cost = {
    'wt': 1150,        # EUR/kW for wind turbines
    'pv': 475,         # EUR/kW for solar PV
    'bat': 250,        # EUR/kWh for batteries
    'elz': 610,        # EUR/kW for electrolyzers
    'plasma': 3000,    # EUR/kW for plasma units
    'methanol': 3500,  # EUR/kW for methanol synthesis
    'h2tank': 400,     # EUR/kg for H2 storage
    'cotank': 400      # EUR/kg for CO storage
}

# OPEX as percentage of CAPEX
opex_factor = {
    'wt': 0.02,
    'pv': 0.02,
    'bat': 0.01,
    'elz': 0.03,
    'plasma': 0.03,
    'methanol': 0.05,
    'h2tank': 0,
    'cotank': 0
}

plant_lifetime = 30    # years
discount_rate = 0.075  # 7.5%

# =============================================================================
# 8. Create the Optimization Model
# =============================================================================
print("Setting up the Gurobi optimization model with typical days...")
start_time = time.time()

# Initialize model
model = gp.Model("Methanol_Production_Typical_Days")

# Step 1: Create decision variables
# Main capacity variables
wt = model.addVar(lb=wt_min, ub=wt_max, name="wind_turbine_capacity")
pv = model.addVar(lb=pv_min, ub=pv_max, name="pv_capacity")
elz = model.addVar(lb=elz_min, ub=elz_max, name="electrolyzer_capacity")
plasma = model.addVar(lb=plasma_min, ub=plasma_max, name="plasma_unit_capacity")
bat = model.addVar(lb=bat_min, ub=bat_max, name="battery_capacity")
h2tank = model.addVar(lb=h2tank_min, ub=h2tank_max, name="h2_tank_capacity")
cotank = model.addVar(lb=cotank_min, ub=cotank_max, name="co_tank_capacity")

# Create variables for each hour in the typical days
P_wt = {}      # Wind power output (kW)
P_pv = {}      # Solar power output (kW)
P_elz = {}     # Power to electrolyzer (kW)
P_plasma = {}  # Power to plasma unit (kW)

H2_prod = {}   # H2 production rate (kg/h)
CO_prod = {}   # CO production rate (kg/h)
H2_supply = {} # H2 supply rate including from storage (kg/h)
CO_supply = {} # CO supply rate including from storage (kg/h)

H2_store = {}  # H2 storage level (kg)
CO_store = {}  # CO storage level (kg)
Bat_store = {} # Battery storage level (kWh)

H2_in = {}     # H2 going into storage (kg/h)
H2_out = {}    # H2 coming out of storage (kg/h)
CO_in = {}     # CO going into storage (kg/h)
CO_out = {}    # CO coming out of storage (kg/h)
Bat_in = {}    # Battery charging (kW)
Bat_out = {}   # Battery discharging (kW)

Deficit = {}   # Binary variable indicating production deficit

# Initialize all time-dependent variables
for t in range(time_periods):
    # Power flows
    P_wt[t] = model.addVar(name=f"P_wt_{t}")
    P_pv[t] = model.addVar(name=f"P_pv_{t}")
    P_elz[t] = model.addVar(lb=0, name=f"P_elz_{t}")
    P_plasma[t] = model.addVar(lb=0, name=f"P_plasma_{t}")
    
    # Production and supply rates
    H2_prod[t] = model.addVar(lb=0, name=f"H2_prod_{t}")
    CO_prod[t] = model.addVar(lb=0, name=f"CO_prod_{t}")
    H2_supply[t] = model.addVar(lb=0, name=f"H2_supply_{t}")
    CO_supply[t] = model.addVar(lb=0, name=f"CO_supply_{t}")
    
    # Storage levels
    H2_store[t] = model.addVar(lb=0, ub=h2tank_max, name=f"H2_store_{t}")
    CO_store[t] = model.addVar(lb=0, ub=cotank_max, name=f"CO_store_{t}")
    Bat_store[t] = model.addVar(lb=bat_min*0.2, ub=bat_max, name=f"Bat_store_{t}")
    
    # Storage flows
    H2_in[t] = model.addVar(lb=0, name=f"H2_in_{t}")
    H2_out[t] = model.addVar(lb=0, name=f"H2_out_{t}")
    CO_in[t] = model.addVar(lb=0, name=f"CO_in_{t}")
    CO_out[t] = model.addVar(lb=0, name=f"CO_out_{t}")
    Bat_in[t] = model.addVar(lb=0, name=f"Bat_in_{t}")
    Bat_out[t] = model.addVar(lb=0, name=f"Bat_out_{t}")
    
    # Deficit indicator (binary variable for each hour)
    Deficit[t] = model.addVar(vtype=GRB.BINARY, name=f"Deficit_{t}")

model.update()

# Step 2: Add constraints
# a) Renewable generation constraints - using the typical day data
for t in range(time_periods):
    model.addConstr(P_wt[t] == wt * typical_wind[t], name=f"wind_gen_{t}")
    model.addConstr(P_pv[t] == pv * typical_solar[t], name=f"pv_gen_{t}")

# b) Power balance constraints
for t in range(time_periods):
    # Total power available = renewables + battery discharge - battery charge - fixed methanol synthesis load
    model.addConstr(
        P_wt[t] + P_pv[t] + Bat_out[t] - Bat_in[t] - p_methanol_installed == 
        P_elz[t] + P_plasma[t], 
        name=f"power_balance_{t}"
    )

# c) Electrolyzer & plasma capacity constraints
for t in range(time_periods):
    model.addConstr(P_elz[t] <= elz, name=f"elz_capacity_{t}")
    model.addConstr(P_plasma[t] <= plasma, name=f"plasma_capacity_{t}")

# d) H2 and CO production constraints
for t in range(time_periods):
    model.addConstr(H2_prod[t] == P_elz[t] / e_elz_h2, name=f"h2_production_{t}")
    model.addConstr(CO_prod[t] == P_plasma[t] / e_plasma_co, name=f"co_production_{t}")

# e) Storage dynamics (special handling for day transitions)
# Initialize storage (start with full storage)
model.addConstr(H2_store[0] == h2tank, name="init_h2_storage")
model.addConstr(CO_store[0] == cotank, name="init_co_storage")
model.addConstr(Bat_store[0] == bat, name="init_battery")

# Storage balance equations for each hour
for t in range(1, time_periods):
    # If we're at the start of a new day (t % 24 == 0), 
    # connect to the end of the previous typical day
    if t % 24 == 0:
        # For day transitions, we need to consider seasonal continuity
        # Storage at the start of a season should be the same as at the end
        prev_t = t - 1
    else:
        prev_t = t - 1
        
    model.addConstr(
        H2_store[t] == H2_store[prev_t] + H2_in[prev_t] - H2_out[prev_t],
        name=f"h2_storage_balance_{t}"
    )
    model.addConstr(
        CO_store[t] == CO_store[prev_t] + CO_in[prev_t] - CO_out[prev_t],
        name=f"co_storage_balance_{t}"
    )
    model.addConstr(
        Bat_store[t] == Bat_store[prev_t] + Bat_in[prev_t]*0.9 - Bat_out[prev_t],
        name=f"battery_balance_{t}"
    )

# Add continuity constraints for each typical day (end of day connects to beginning of next day)
for day_idx in range(len(typical_days) - 1):
    day_end = (day_idx + 1) * 24 - 1
    next_day_start = (day_idx + 1) * 24
    
    # Storage level at the end of one day must match the beginning of the next day
    model.addConstr(
        H2_store[next_day_start] == H2_store[day_end] + H2_in[day_end] - H2_out[day_end],
        name=f"h2_day_transition_{day_idx}"
    )
    model.addConstr(
        CO_store[next_day_start] == CO_store[day_end] + CO_in[day_end] - CO_out[day_end],
        name=f"co_day_transition_{day_idx}"
    )
    model.addConstr(
        Bat_store[next_day_start] == Bat_store[day_end] + Bat_in[day_end]*0.9 - Bat_out[day_end],
        name=f"bat_day_transition_{day_idx}"
    )

# Ensure seasonal continuity (last hour of last day connects to first hour of first day)
last_hour = time_periods - 1
model.addConstr(
    H2_store[0] == H2_store[last_hour] + H2_in[last_hour] - H2_out[last_hour],
    name="h2_seasonal_continuity"
)
model.addConstr(
    CO_store[0] == CO_store[last_hour] + CO_in[last_hour] - CO_out[last_hour],
    name="co_seasonal_continuity"
)
model.addConstr(
    Bat_store[0] == Bat_store[last_hour] + Bat_in[last_hour]*0.9 - Bat_out[last_hour],
    name="bat_seasonal_continuity"
)

# f) Supply-demand balance and deficit indicator
for t in range(time_periods):
    # H2 and CO supply includes both production and storage outflow
    model.addConstr(H2_supply[t] == H2_prod[t] + H2_out[t], name=f"h2_supply_{t}")
    model.addConstr(CO_supply[t] == CO_prod[t] + CO_out[t], name=f"co_supply_{t}")
    
    # Storage inflow is excess production
    model.addConstr(H2_in[t] <= H2_prod[t], name=f"h2_storage_in_{t}")
    model.addConstr(CO_in[t] <= CO_prod[t], name=f"co_storage_in_{t}")
    
    # Ensure that production + storage withdrawal >= demand - deficit
    model.addConstr(
        H2_supply[t] >= m_h2_req_hour - m_h2_req_hour * Deficit[t],
        name=f"h2_deficit_{t}"
    )
    model.addConstr(
        CO_supply[t] >= m_co_req_hour - m_co_req_hour * Deficit[t],
        name=f"co_deficit_{t}"
    )

# g) Total acceptable deficit constraint (scaled for typical days)
model.addConstr(
    gp.quicksum(Deficit[t] for t in range(time_periods)) * annual_scaling_factor / time_periods <= TAD_threshold,
    name="total_deficit_constraint"
)

# Step 3: Set up the objective function
# Calculate CAPEX
capex_wt = unit_cost['wt'] * wt
capex_pv = unit_cost['pv'] * pv
capex_elz = unit_cost['elz'] * elz
capex_plasma = unit_cost['plasma'] * plasma
capex_methanol = unit_cost['methanol'] * p_methanol_installed
capex_bat = unit_cost['bat'] * bat
capex_h2tank = unit_cost['h2tank'] * h2tank
capex_cotank = unit_cost['cotank'] * cotank

total_capex = capex_wt + capex_pv + capex_elz + capex_plasma + capex_methanol + capex_bat + capex_h2tank + capex_cotank

# Calculate OPEX
opex_wt = opex_factor['wt'] * capex_wt
opex_pv = opex_factor['pv'] * capex_pv
opex_elz = opex_factor['elz'] * capex_elz
opex_plasma = opex_factor['plasma'] * capex_plasma
opex_methanol = opex_factor['methanol'] * capex_methanol
opex_bat = opex_factor['bat'] * capex_bat
opex_h2tank = opex_factor['h2tank'] * capex_h2tank
opex_cotank = opex_factor['cotank'] * capex_cotank

total_opex = opex_wt + opex_pv + opex_elz + opex_plasma + opex_methanol + opex_bat + opex_h2tank + opex_cotank

# Calculate annualized capital cost
annualization_factor = (discount_rate * (1 + discount_rate)**plant_lifetime) / ((1 + discount_rate)**plant_lifetime - 1)
annual_capex = total_capex * annualization_factor

# Total annual cost
total_annual_cost = annual_capex + total_opex

# LCOC calculation (EUR/ton methanol)
# For the objective, we'll use tonnes_methanol_prod_year_real which is more realistic
lcoc = total_annual_cost / tonnes_methanol_prod_year_real

# Set the objective
model.setObjective(lcoc, GRB.MINIMIZE)

# =============================================================================
# 9. Solve the Model
# =============================================================================
print("Starting optimization...")
model.optimize()

# Check if a solution was found
if model.status == GRB.OPTIMAL:
    print("\nOptimal solution found!")
    
    # Extract the optimal values
    optimal_wt = wt.X
    optimal_pv = pv.X
    optimal_elz = elz.X
    optimal_plasma = plasma.X
    optimal_bat = bat.X
    optimal_h2tank = h2tank.X
    optimal_cotank = cotank.X
    
    # Calculate the optimal LCOC
    optimal_lcoc = model.objVal
    
    # Calculate total deficit hours (scaled to annual)
    typical_day_deficit = sum(Deficit[t].X for t in range(time_periods))
    annual_deficit_hours = typical_day_deficit * annual_scaling_factor / time_periods
    
    # Display the results
    print("\n=== OPTIMIZATION RESULTS ===")
    print(f"Wind Turbine Capacity: {optimal_wt:.2f} kW")
    print(f"Solar PV Capacity: {optimal_pv:.2f} kW")
    print(f"Electrolyzer Capacity: {optimal_elz:.2f} kW")
    print(f"Plasma Unit Capacity: {optimal_plasma:.2f} kW")
    print(f"Battery Capacity: {optimal_bat:.2f} kWh")
    print(f"H2 Storage Capacity: {optimal_h2tank:.2f} kg")
    print(f"CO Storage Capacity: {optimal_cotank:.2f} kg")
    print(f"LCOC: {optimal_lcoc:.2f} EUR/ton CH3OH")
    print(f"Deficit Hours (typical days): {typical_day_deficit}")
    print(f"Projected Annual Deficit Hours: {annual_deficit_hours:.1f}")
    
    # Calculate additional performance metrics
    total_capex_value = (unit_cost['wt'] * optimal_wt + 
                         unit_cost['pv'] * optimal_pv + 
                         unit_cost['elz'] * optimal_elz + 
                         unit_cost['plasma'] * optimal_plasma + 
                         unit_cost['methanol'] * p_methanol_installed + 
                         unit_cost['bat'] * optimal_bat + 
                         unit_cost['h2tank'] * optimal_h2tank + 
                         unit_cost['cotank'] * optimal_cotank)
    
    print(f"Total CAPEX: {total_capex_value/1e6:.2f} MEUR")
    
    # Extract time series results for plotting
    time_data = {
        'Hour': list(range(time_periods)),
        'OriginalHour': hour_mapping,
        'Season': ['Winter'] * 24 + ['Spring'] * 24 + ['Summer'] * 24 + ['Fall'] * 24,
        'Day': [f"Day {day}" for day in typical_days for _ in range(24)],
        'HourOfDay': list(range(24)) * 4,
        'Wind_Power': [P_wt[t].X for t in range(time_periods)],
        'Solar_Power': [P_pv[t].X for t in range(time_periods)],
        'Electrolyzer_Power': [P_elz[t].X for t in range(time_periods)],
        'Plasma_Power': [P_plasma[t].X for t in range(time_periods)],
        'H2_Production': [H2_prod[t].X for t in range(time_periods)],
        'CO_Production': [CO_prod[t].X for t in range(time_periods)],
        'H2_Storage': [H2_store[t].X for t in range(time_periods)],
        'CO_Storage': [CO_store[t].X for t in range(time_periods)],
        'Battery_Level': [Bat_store[t].X for t in range(time_periods)],
        'Deficit': [Deficit[t].X for t in range(time_periods)]
    }
    
    # Create a DataFrame for easier analysis
    results_df = pd.DataFrame(time_data)
    
    # Calculate renewable generation for typical days
    typical_renewable_generation = sum(P_wt[t].X + P_pv[t].X for t in range(time_periods))
    # Scale to annual values
    annual_renewable_generation = typical_renewable_generation * annual_scaling_factor / time_periods
    print(f"Projected Annual Renewable Generation: {annual_renewable_generation:.2f} kWh")
    
    # Calculate H2 and CO production for typical days
    h2_produced_typical = sum(H2_prod[t].X for t in range(time_periods))
    co_produced_typical = sum(CO_prod[t].X for t in range(time_periods))
    # Scale to annual values
    h2_produced_annual = h2_produced_typical * annual_scaling_factor / time_periods
    co_produced_annual = co_produced_typical * annual_scaling_factor / time_periods
    print(f"Projected Annual H2 Produced: {h2_produced_annual:.2f} kg")
    print(f"Projected Annual CO Produced: {co_produced_annual:.2f} kg")
    
    # Calculate battery and tank cycles for typical days
    bat_cycles_typical = sum(Bat_in[t].X for t in range(time_periods)) / optimal_bat if optimal_bat > 0 else 0
    h2_tank_cycles_typical = sum(H2_in[t].X for t in range(time_periods)) / optimal_h2tank if optimal_h2tank > 0 else 0
    co_tank_cycles_typical = sum(CO_in[t].X for t in range(time_periods)) / optimal_cotank if optimal_cotank > 0 else 0
    # Scale to annual values
    bat_cycles_annual = bat_cycles_typical * annual_scaling_factor / time_periods
    h2_tank_cycles_annual = h2_tank_cycles_typical * annual_scaling_factor / time_periods
    co_tank_cycles_annual = co_tank_cycles_typical * annual_scaling_factor / time_periods
    print(f"Projected Annual Battery Cycles: {bat_cycles_annual:.2f}")
    print(f"Projected Annual H2 Tank Cycles: {h2_tank_cycles_annual:.2f}")
    print(f"Projected Annual CO Tank Cycles: {co_tank_cycles_annual:.2f}")
    
    # Save the results to a CSV file
    output_data = {
        'Parameter': [
            'wind_turbine_capacity', 'solar_pv_capacity', 
            'electrolyzer_capacity', 'plasma_unit_capacity',
            'battery_capacity', 'h2_tank_capacity', 'co_tank_capacity',
            'lcoc', 'deficit_hours_typical', 'deficit_hours_annual', 'total_capex',
            'renewable_generation_annual', 'h2_produced_annual', 'co_produced_annual',
            'battery_cycles_annual', 'h2_tank_cycles_annual', 'co_tank_cycles_annual'
        ],
        'Value': [
            optimal_wt, optimal_pv, 
            optimal_elz, optimal_plasma,
            optimal_bat, optimal_h2tank, optimal_cotank,
            optimal_lcoc, typical_day_deficit, annual_deficit_hours, total_capex_value,
            annual_renewable_generation, h2_produced_annual, co_produced_annual,
            bat_cycles_annual, h2_tank_cycles_annual, co_tank_cycles_annual
        ]
    }
    
    with open('gurobi_typical_days_results.csv', 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['Parameter', 'Value'])
        for i in range(len(output_data['Parameter'])):
            writer.writerow([output_data['Parameter'][i], output_data['Value'][i]])
    
    # =============================================================================
    # 10. Visualize the Results
    # =============================================================================
    # Set up the plotting style
    plt.style.use('seaborn-v0_8-darkgrid')
    
    # 1. Capacity mix pie chart
    plt.figure(figsize=(12, 8))
    capacity_labels = ['Wind', 'Solar PV', 'Electrolyzer', 'Plasma Unit', 'Battery']
    capacity_values = [optimal_wt, optimal_pv, optimal_elz, optimal_plasma, optimal_bat]
    capacity_colors = ['skyblue', 'gold', 'lightgreen', 'tomato', 'mediumpurple']
    
    plt.pie(capacity_values, labels=capacity_labels, autopct='%1.1f%%', 
            colors=capacity_colors, startangle=90)
    plt.title('Optimal Capacity Mix', fontsize=16)
    plt.savefig('typical_days_capacity_mix.png', dpi=300, bbox_inches='tight')
    
    # 2. CAPEX breakdown
    plt.figure(figsize=(12, 8))
    capex_labels = ['Wind', 'Solar PV', 'Electrolyzer', 'Plasma Unit', 'Methanol Synth', 'Battery', 'H2 Tank', 'CO Tank']
    capex_values = [
        unit_cost['wt'] * optimal_wt / 1e6,
        unit_cost['pv'] * optimal_pv / 1e6,
        unit_cost['elz'] * optimal_elz / 1e6,
        unit_cost['plasma'] * optimal_plasma / 1e6,
        unit_cost['methanol'] * p_methanol_installed / 1e6,
        unit_cost['bat'] * optimal_bat / 1e6,
        unit_cost['h2tank'] * optimal_h2tank / 1e6,
        unit_cost['cotank'] * optimal_cotank / 1e6
    ]
    
    plt.barh(capex_labels, capex_values, color=plt.cm.tab10.colors)
    plt.xlabel('CAPEX (Million EUR)', fontsize=14)
    plt.title('CAPEX Breakdown by Component', fontsize=16)
    plt.grid(alpha=0.3)
    for i, v in enumerate(capex_values):
        plt.text(v + 0.1, i, f'{v:.1f}', va='center')
    plt.savefig('typical_days_capex_breakdown.png', dpi=300, bbox_inches='tight')
    
    # 3. Time series for all typical days (4 days  24 hours)
    plt.figure(figsize=(16, 12))
    
    # Create a subplot for each season
    for day_idx, season in enumerate(['Winter', 'Spring', 'Summer', 'Fall']):
        start_hour = day_idx * 24
        end_hour = start_hour + 24
        
        # Power generation subplot
        plt.subplot(4, 3, day_idx*3 + 1)
        plt.plot(results_df['HourOfDay'][start_hour:end_hour], 
                 results_df['Wind_Power'][start_hour:end_hour], 
                 label='Wind', color='skyblue')
        plt.plot(results_df['HourOfDay'][start_hour:end_hour], 
                 results_df['Solar_Power'][start_hour:end_hour], 
                 label='Solar', color='gold')
        if day_idx == 0:
            plt.title('Renewable Power Generation', fontsize=14)
        plt.ylabel(f'{season} (kW)', fontsize=12)
        if day_idx == 3:  # Only add x-label for the bottom plot
            plt.xlabel('Hour of Day', fontsize=12)
        if day_idx == 0:  # Only add legend for the top plot
            plt.legend(loc='upper right')
        plt.grid(alpha=0.3)
        
        # Production rates subplot
        plt.subplot(4, 3, day_idx*3 + 2)
        plt.plot(results_df['HourOfDay'][start_hour:end_hour], 
                 results_df['H2_Production'][start_hour:end_hour], 
                 label='H2 Prod', color='lightgreen')
        plt.plot(results_df['HourOfDay'][start_hour:end_hour], 
                 results_df['CO_Production'][start_hour:end_hour], 
                 label='CO Prod', color='tomato')
        plt.axhline(y=m_h2_req_hour, linestyle='--', color='green', alpha=0.5, label='H2 Target')
        plt.axhline(y=m_co_req_hour, linestyle='--', color='red', alpha=0.5, label='CO Target')
        if day_idx == 0:
            plt.title('Production Rates', fontsize=14)
        if day_idx == 3:  # Only add x-label for the bottom plot
            plt.xlabel('Hour of Day', fontsize=12)
        if day_idx == 0:  # Only add legend for the top plot
            plt.legend(loc='upper right')
        plt.grid(alpha=0.3)
        
        # Storage levels subplot
        plt.subplot(4, 3, day_idx*3 + 3)
        plt.plot(results_df['HourOfDay'][start_hour:end_hour], 
                 results_df['H2_Storage'][start_hour:end_hour], 
                 label='H2 Storage', color='green')
        plt.plot(results_df['HourOfDay'][start_hour:end_hour], 
                 results_df['CO_Storage'][start_hour:end_hour], 
                 label='CO Storage', color='red')
        plt.plot(results_df['HourOfDay'][start_hour:end_hour], 
                 results_df['Battery_Level'][start_hour:end_hour], 
                 label='Battery', color='purple')
        if day_idx == 0:
            plt.title('Storage Levels', fontsize=14)
        if day_idx == 3:  # Only add x-label for the bottom plot
            plt.xlabel('Hour of Day', fontsize=12)
        if day_idx == 0:  # Only add legend for the top plot
            plt.legend(loc='upper right')
        plt.grid(alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('typical_days_time_series.png', dpi=300, bbox_inches='tight')
    
    # 4. Seasonal comparison of key metrics
    plt.figure(figsize=(14, 10))
    
    # Group data by season
    seasonal_data = {
        'Season': ['Winter', 'Spring', 'Summer', 'Fall'],
        'Avg_Wind': [],
        'Avg_Solar': [],
        'Avg_H2_Prod': [],
        'Avg_CO_Prod': [],
        'Deficit_Hours': []
    }
    
    for day_idx, season in enumerate(['Winter', 'Spring', 'Summer', 'Fall']):
        start_hour = day_idx * 24
        end_hour = start_hour + 24
        
        seasonal_data['Avg_Wind'].append(results_df['Wind_Power'][start_hour:end_hour].mean())
        seasonal_data['Avg_Solar'].append(results_df['Solar_Power'][start_hour:end_hour].mean())
        seasonal_data['Avg_H2_Prod'].append(results_df['H2_Production'][start_hour:end_hour].mean())
        seasonal_data['Avg_CO_Prod'].append(results_df['CO_Production'][start_hour:end_hour].mean())
        seasonal_data['Deficit_Hours'].append(results_df['Deficit'][start_hour:end_hour].sum())
    
    # Create subplots for seasonal comparisons
    plt.subplot(2, 2, 1)
    plt.bar(seasonal_data['Season'], seasonal_data['Avg_Wind'], color='skyblue')
    plt.title('Average Wind Power by Season', fontsize=14)
    plt.ylabel('Power (kW)', fontsize=12)
    plt.grid(axis='y', alpha=0.3)
    
    plt.subplot(2, 2, 2)
    plt.bar(seasonal_data['Season'], seasonal_data['Avg_Solar'], color='gold')
    plt.title('Average Solar Power by Season', fontsize=14)
    plt.ylabel('Power (kW)', fontsize=12)
    plt.grid(axis='y', alpha=0.3)
    
    plt.subplot(2, 2, 3)
    plt.bar(seasonal_data['Season'], 
            [seasonal_data['Avg_H2_Prod'][i] + seasonal_data['Avg_CO_Prod'][i] 
             for i in range(4)], color='lightgreen')
    plt.title('Average Total Production by Season', fontsize=14)
    plt.ylabel('Production (kg/h)', fontsize=12)
    plt.grid(axis='y', alpha=0.3)
    
    plt.subplot(2, 2, 4)
    plt.bar(seasonal_data['Season'], seasonal_data['Deficit_Hours'], color='indianred')
    plt.title('Deficit Hours by Season', fontsize=14)
    plt.ylabel('Hours', fontsize=12)
    plt.grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('typical_days_seasonal_comparison.png', dpi=300, bbox_inches='tight')
    
    # 5. Daily profiles by season (renewable generation, production, storage)
    plt.figure(figsize=(15, 18))
    
    # Create profile plots for each metric, with all seasons overlaid
    metrics = [
        ('Wind_Power', 'Wind Power', 'skyblue', 'kW'),
        ('Solar_Power', 'Solar Power', 'gold', 'kW'),
        ('H2_Production', 'H2 Production', 'lightgreen', 'kg/h'),
        ('CO_Production', 'CO Production', 'tomato', 'kg/h'),
        ('H2_Storage', 'H2 Storage', 'green', 'kg'),
        ('CO_Storage', 'CO Storage', 'red', 'kg')
    ]
    
    for i, (metric, title, color, unit) in enumerate(metrics):
        plt.subplot(3, 2, i+1)
        
        for day_idx, season in enumerate(['Winter', 'Spring', 'Summer', 'Fall']):
            start_hour = day_idx * 24
            end_hour = start_hour + 24
            
            plt.plot(results_df['HourOfDay'][start_hour:end_hour], 
                     results_df[metric][start_hour:end_hour], 
                     label=season, 
                     linewidth=2)
        
        plt.title(f'Daily {title} Profile by Season', fontsize=14)
        plt.xlabel('Hour of Day', fontsize=12)
        plt.ylabel(unit, fontsize=12)
        plt.legend()
        plt.grid(alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('typical_days_daily_profiles.png', dpi=300, bbox_inches='tight')
    
    # 6. Deficit pattern visualization
    plt.figure(figsize=(14, 6))
    
    # Create a heatmap-style visualization of deficits
    deficit_matrix = np.zeros((4, 24))
    
    for day_idx in range(4):
        start_hour = day_idx * 24
        end_hour = start_hour + 24
        
        for h in range(24):
            deficit_matrix[day_idx, h] = results_df['Deficit'][start_hour + h]
    
    plt.imshow(deficit_matrix, cmap='Reds', aspect='auto')
    plt.colorbar(label='Deficit (1=Yes, 0=No)')
    plt.yticks(range(4), ['Winter', 'Spring', 'Summer', 'Fall'])
    plt.xticks(range(0, 24, 2))
    plt.xlabel('Hour of Day', fontsize=14)
    plt.ylabel('Season', fontsize=14)
    plt.title('Production Deficit Pattern by Season and Hour', fontsize=16)
    plt.grid(False)
    
    plt.tight_layout()
    plt.savefig('typical_days_deficit_pattern.png', dpi=300, bbox_inches='tight')
    
    print(f"\nOptimization completed in {time.time() - start_time:.2f} seconds")
    print("Results saved to 'gurobi_typical_days_results.csv'")
    print("Visualizations saved as PNG files")

else:
    print(f"Optimization failed with status code {model.status}")
```
