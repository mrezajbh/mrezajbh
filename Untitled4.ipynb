```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.integrate import simps
from scipy.integrate import quad
from scipy import stats



class COAdsorptionModel:
    def __init__(self):
        self.R = 8.314  # Universal gas constant (J/(mol·K))
        self.isotherm_data = {
            303: {'p': np.array([0, 50, 100, 200, 300, 400, 500]),
                  'q': np.array([0, 2.45, 2.72, 2.90, 2.98, 3.02, 3.05])},
            313: {'p': np.array([0, 50, 100, 200, 300, 400, 500]),
                  'q': np.array([0, 2.30, 2.55, 2.75, 2.85, 2.90, 2.95])},
            323: {'p': np.array([0, 50, 100, 200, 300, 400, 500]),
                  'q': np.array([0, 2.15, 2.40, 2.60, 2.70, 2.75, 2.80])},
            333: {'p': np.array([0, 50, 100, 200, 300, 400, 500]),
                  'q': np.array([0, 1.95, 2.20, 2.40, 2.50, 2.55, 2.60])}
        }
        self.langmuir_params = {
            303: {'qm': 2.99, 'b': 0.0979},
            313: {'qm': 2.84, 'b': 0.0860},
            323: {'qm': 2.66, 'b': 0.0792},
            333: {'qm': 2.56, 'b': 0.0580}
        }
        self.sips_params = {
            303: {'qm': 2.99, 'b': 0.0961, 'n': 0.993},
            313: {'qm': 2.78, 'b': 0.0562, 'n': 0.859},
            323: {'qm': 2.60, 'b': 0.0479, 'n': 0.840},
            333: {'qm': 2.49, 'b': 0.0352, 'n': 0.848}
        }

    def langmuir_equation(self, p, qm, b):
        return qm * b * p / (1 + b * p)

    def sips_equation(self, p, qm, b, n):
        return qm * (b * p**n) / (1 + b * p**n)

    def plot_isotherm(self, T, model='langmuir'):
        p = self.isotherm_data[T]['p']
        q = self.isotherm_data[T]['q']
        
        plt.figure(figsize=(10, 6))
        plt.scatter(p, q, label='Experimental Data')
        
        p_smooth = np.linspace(0, max(p), 100)
        if model == 'langmuir':
            params = self.langmuir_params[T]
            q_fit = self.langmuir_equation(p_smooth, params['qm'], params['b'])
            plt.plot(p_smooth, q_fit, 'r-', label='Langmuir Fit')
        elif model == 'sips':
            params = self.sips_params[T]
            q_fit = self.sips_equation(p_smooth, params['qm'], params['b'], params['n'])
            plt.plot(p_smooth, q_fit, 'r-', label='Sips Fit')
        
        plt.xlabel('Pressure (kPa)')
        plt.ylabel('Adsorbed Amount (mmol/g)')
        plt.title(f'CO Adsorption on CuCl/Y at {T} K')
        plt.legend()
        plt.grid(True)
        plt.show()

    def plot_all_isotherms(self, model='langmuir'):
        plt.figure(figsize=(12, 8))
        for T in self.isotherm_data.keys():
            p = self.isotherm_data[T]['p']
            q = self.isotherm_data[T]['q']
            plt.scatter(p, q, label=f'{T} K (Experimental)')
            
            p_smooth = np.linspace(0, max(p), 100)
            if model == 'langmuir':
                params = self.langmuir_params[T]
                q_fit = self.langmuir_equation(p_smooth, params['qm'], params['b'])
            elif model == 'sips':
                params = self.sips_params[T]
                q_fit = self.sips_equation(p_smooth, params['qm'], params['b'], params['n'])
            plt.plot(p_smooth, q_fit, label=f'{T} K ({model.capitalize()} Fit)')
        
        plt.xlabel('Pressure (kPa)')
        plt.ylabel('Adsorbed Amount (mmol/g)')
        plt.title(f'CO Adsorption on CuCl/Y at Different Temperatures ({model.capitalize()} Model)')
        plt.legend()
        plt.grid(True)
        plt.show()

        

    def calculate_surface_potential(self, T, p):
        T_values = list(self.langmuir_params.keys())
        if T in T_values:
            params = self.langmuir_params[T]
        elif min(T_values) < T < max(T_values):  # Ensure T is within the data range
            # Interpolate parameters
            T_low = max([t for t in T_values if t < T])
            T_high = min([t for t in T_values if t > T])
            params_low = self.langmuir_params[T_low]
            params_high = self.langmuir_params[T_high]
            w = (T - T_low) / (T_high - T_low)
            params = {
                'qm': params_low['qm'] * (1 - w) + params_high['qm'] * w,
                'b': params_low['b'] * (1 - w) + params_high['b'] * w
            }
        else:
            # Handle out-of-range T (e.g., raise an error or extrapolate)
            raise ValueError(f"Temperature {T} K is outside the range of available data.")
    
        qm, b = params['qm'], params['b']
        return -self.R * T * np.log(1 + b * p)

    
    def calculate_delta_G(self, T, p):
        params = self.langmuir_params[T]
        qm, b = params['qm'], params['b']
        q = self.langmuir_equation(p, qm, b)
        X = self.calculate_surface_potential(T, p)
        return X / q

    def calculate_delta_H(self, T, p):
        dT = 0.1  # Small temperature difference
        X_T = self.calculate_surface_potential(T, p)
    
        # Check if T + dT is within the valid range before calculating X_T_plus
        if T + dT <= max(self.langmuir_params.keys()):
            X_T_plus = self.calculate_surface_potential(T + dT, p)
        else:
            # Handle the case where T + dT is out of range (e.g., use the highest available T)
            X_T_plus = self.calculate_surface_potential(max(self.langmuir_params.keys()), p) 
    
        return -T**2 * (X_T_plus / (T + dT) - X_T / T) / dT

    def calculate_delta_S(self, T, p):
        T_values = np.linspace(min(self.langmuir_params.keys()), max(self.langmuir_params.keys()), 100)  # Generate a range of T values
        X_values = [self.calculate_surface_potential(t, p) for t in T_values]
        return -np.gradient(X_values, T_values)[np.where(T_values == T)[0][0]]  # Calculate gradient and extract value at the specific T


    def plot_thermodynamic_properties(self):
        p_values = np.linspace(1, 300, 100)  # kPa

        # Plot Delta G
        plt.figure(figsize=(10, 6))
        for T in self.langmuir_params.keys():
            delta_G_values = [self.calculate_delta_G(T, p) / 1000 for p in p_values]  # Convert to kJ/mol
            plt.plot(p_values, delta_G_values, label=f'{T} K', marker='o', linestyle='-')  # Add markers and lines
        plt.xlabel('Pressure (kPa)', fontsize=12)  # Increase font size
        plt.ylabel('ΔG (kJ/mol)', fontsize=12)
        plt.title('Gibbs Free Energy Change of CO Adsorption on CuCl/Y', fontsize=14)  # Increase title font size
        plt.legend(fontsize=10)  # Adjust legend font size
        plt.grid(True)
        plt.tight_layout()  # Prevent overlapping elements
        plt.show()

        # Plot Delta H
        plt.figure(figsize=(10, 6))
        for T in self.langmuir_params.keys():
            delta_H_values = [self.calculate_delta_H(T, p) / 1000 for p in p_values]  # Convert to kJ/mol
            plt.plot(p_values, delta_H_values, label=f'{T} K', marker='s', linestyle='--')  # Different markers and lines
        plt.xlabel('Pressure (kPa)', fontsize=12)
        plt.ylabel('ΔH (kJ/mol)', fontsize=12)
        plt.title('Enthalpy Change of CO Adsorption on CuCl/Y', fontsize=14)
        plt.legend(fontsize=10)
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        # Plot T * Delta S
        plt.figure(figsize=(10, 6))
        for T in self.langmuir_params.keys():
            delta_S_values = [T * self.calculate_delta_S(T, p) / 1000 for p in p_values]  # Convert to kJ/mol
            plt.plot(p_values, delta_S_values, label=f'{T} K', marker='^', linestyle='-.')  # More marker/line variety
        plt.xlabel('Pressure (kPa)', fontsize=12)
        plt.ylabel('TΔS (kJ/mol)', fontsize=12)
        plt.title('Entropy Change of CO Adsorption on CuCl/Y', fontsize=14)
        plt.legend(fontsize=10)
        plt.grid(True)
        plt.tight_layout()
        plt.show()


    def plot_parameter_variation(self):
        T_values = list(self.langmuir_params.keys())
    
        # Langmuir Parameters
        qm_langmuir = [self.langmuir_params[T]['qm'] for T in T_values]
        b_langmuir = [self.langmuir_params[T]['b'] for T in T_values]
    
        plt.figure(figsize=(10, 6))
        plt.plot(T_values, qm_langmuir, 'o-', label='qm (Langmuir)')
        plt.plot(T_values, b_langmuir, 's-', label='b (Langmuir)')
        plt.xlabel('Temperature (K)')
        plt.ylabel('Parameter Value')
        plt.title('Langmuir Parameter Variation with Temperature')
        plt.legend()
        plt.grid(True)
        plt.show()
    
        # Sips Parameters
        qm_sips = [self.sips_params[T]['qm'] for T in T_values]
        b_sips = [self.sips_params[T]['b'] for T in T_values]
        n_sips = [self.sips_params[T]['n'] for T in T_values]
    
        plt.figure(figsize=(10, 6))
        plt.plot(T_values, qm_sips, 'o-', label='qm (Sips)')
        plt.plot(T_values, b_sips, 's-', label='b (Sips)')
        plt.plot(T_values, n_sips, '^-' ,label='n (Sips)')
        plt.xlabel('Temperature (K)')
        plt.ylabel('Parameter Value')
        plt.title('Sips Parameter Variation with Temperature')
        plt.legend()
        plt.grid(True)
        plt.show()

    def plot_surface_potential_vs_pressure(self):
        p_values = np.linspace(1, 300, 100)  # kPa
    
        plt.figure(figsize=(10, 6))
        for T in self.langmuir_params.keys():
            X_values = [self.calculate_surface_potential(T, p) / 1000 for p in p_values]  # Convert to kJ/mol
            plt.plot(p_values, X_values, label=f'{T} K')
        plt.xlabel('Pressure (kPa)')
        plt.ylabel('Surface Potential (kJ/mol)')
        plt.title('Surface Potential vs. Pressure at Different Temperatures')
        plt.legend()
        plt.grid(True)
        plt.show()

    def plot_adsorbed_amount_vs_temperature(self, p_constant):
        T_values = list(self.langmuir_params.keys())
    
        plt.figure(figsize=(10, 6))
        for model_name, params_dict in [('Langmuir', self.langmuir_params), ('Sips', self.sips_params)]:
            q_values = []
            for T in T_values:
                params = params_dict[T]
                if model_name == 'Langmuir':
                    q = self.langmuir_equation(p_constant, params['qm'], params['b'])
                else:  # Sips
                    q = self.sips_equation(p_constant, params['qm'], params['b'], params['n'])
                q_values.append(q)
            plt.plot(T_values, q_values, label=f'{model_name} at {p_constant} kPa')
    
        plt.xlabel('Temperature (K)')
        plt.ylabel('Adsorbed Amount (mmol/g)')
        plt.title(f'Adsorbed Amount vs. Temperature at Constant Pressure ({p_constant} kPa)')
        plt.legend()
        plt.grid(True)
        plt.show()



# Usage
model = COAdsorptionModel()

# Plot isotherms and fits
model.plot_all_isotherms()
for T in model.isotherm_data.keys():
    model.plot_isotherm(T, model='langmuir')
    model.plot_isotherm(T, model='sips')

# Plot thermodynamic properties
model.plot_thermodynamic_properties()

# Plot additional visualizations
model.plot_parameter_variation()
model.plot_surface_potential_vs_pressure()
model.plot_adsorbed_amount_vs_temperature(p_constant=100)  # Choose a suitable constant pressure
```
